````python
# Step 1: ê³„ì•½ ë°ì´í„° ì¡°íšŒ (ì‹ ê³„ì•½ ë§ˆìŠ¤í„° + ìƒì„¸ + í•´ì•½ ì •ë³´)
# ê³ ê°ì½”ë“œ ë¦¬ìŠ¤íŠ¸ ìƒì„±
custcode_list = ','.join([f"'{code}'" for code in final_df['ê³ ê°ì½”ë“œ'].unique()])

# ê³„ì•½ ì •ë³´ ì¡°íšŒ ì¿¼ë¦¬ (í•´ì•½ì—¬ë¶€ê´€ë ¨_ì •ë¦¬_v0_2.ipynbì˜ ë¡œì§ ì°¸ì¡°)
contract_query = f"""
SELECT 
    A.ê³ ê°ì½”ë“œ,
    A.ê³„ì•½ì¼ë ¨ë²ˆí˜¸,
    A.ê³„ì•½ì¼ì,
    B.ê³„ì•½ëŒ€ìƒ,
    B.ì‹œì‘ì¼ì,
    B.ì¢…ë£Œì¼ì,
    B.í•´ì•½ì¼ë ¨ë²ˆí˜¸,
    Cust.ê³ ê°ëª… as ê³ ê°ëª…_CU,
    Cust.ê³ ê°êµ¬ë¶„ as ê³ ê°êµ¬ë¶„_CU,
    D.í•´ì•½ì¼ì as í•´ì•½ì¼ì_SM,
    m.Gerldesp,
    m.CONTDIVS as ê³„ì•½ëŒ€ìƒ_ì¤‘ë¶„ë¥˜,
    m.M_nm as ê³„ì•½ëŒ€ìƒ_ì¤‘ë¶„ë¥˜ëª…,
    m.CONTLATY as ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜,
    m.L_nm as ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜ëª…
FROM CESCOEIS.dbo.TB_ì‹ ê³„ì•½_ë§ˆìŠ¤íƒ€ A
LEFT JOIN CESCOEIS.dbo.TB_ì‹ ê³„ì•½_ìƒì„¸ B
ON A.ê³„ì•½ì¼ë ¨ë²ˆí˜¸ = B.ê³„ì•½ì¼ë ¨ë²ˆí˜¸
LEFT JOIN CESCOEIS.dbo.TB_ì‹ ê³„ì•½_ë§ˆìŠ¤íƒ€_í•´ì•½ C
ON A.ê³„ì•½ì¼ë ¨ë²ˆí˜¸ = C.ê³„ì•½ì¼ë ¨ë²ˆí˜¸
LEFT JOIN CESCOEIS.dbo.TB_ì‹ ê³„ì•½_ë¶€ë¶„í•´ì•½ D
ON B.í•´ì•½ì¼ë ¨ë²ˆí˜¸ = D.í•´ì•½ì¼ë ¨ë²ˆí˜¸
LEFT JOIN (
    SELECT ê³ ê°ì½”ë“œ, ê³ ê°ëª…, ê³ ê°êµ¬ë¶„, ë³¸ì‚¬ì½”ë“œ, ë“±ë¡ì¼ì FROM CESCOEIS.dbo.TB_ê³ ê°
    UNION ALL
    SELECT ê³ ê°ì½”ë“œ, ê³ ê°ëª…, ê³ ê°êµ¬ë¶„, ë³¸ì‚¬ì½”ë“œ, ë“±ë¡ì¼ì FROM CESCOEIS.dbo.TB_ë‹¨ë…ê³ ê°
) Cust
ON A.ê³ ê°ì½”ë“œ = Cust.ê³ ê°ì½”ë“œ
LEFT JOIN (
    SELECT MK.Gerlcode, MK.Gerlgubn, MK.Gerldesp
            , CD.CONTDIVS, M.NAME as M_nm
            , CD.CONTLATY, L.NAME as L_nm
    FROM CESCOEIS.dbo.MKTCGERL MK
    LEFT OUTER JOIN CESCOMIS.dbo.TB_CP_CONTRACTDETAIL_HIS CD
    ON MK.Gerlcode = CD.CONTTARG
    AND LEFT('20250501',6) BETWEEN CD.ST_YYYYMM AND CD.ED_YYYYMM
    LEFT OUTER JOIN CESCOMIS.dbo.TB_CODE L
    ON CD.CONTLATY = L.CODE
    LEFT OUTER JOIN CESCOMIS.dbo.TB_CODE M
    ON CD.CONTDIVS = M.CODE
    WHERE Gerlgubn IN ('ê³„ì•½ëŒ€ìƒ','ê°€ì •ì§‘ê³„ì•½ëŒ€ìƒ')
) m
ON B.ê³„ì•½ëŒ€ìƒ = m.Gerlcode
WHERE B.ê³„ì•½ì¼ë ¨ë²ˆí˜¸ IS NOT NULL
AND A.ê³ ê°ì½”ë“œ IN ({custcode_list})
AND A.ê³ ê°ì½”ë“œ IS NOT NULL
AND A.ê³ ê°ì½”ë“œ != ''
AND A.ê³ ê°ì½”ë“œ != 'AT7728'
"""

print("Fetching contract data with joins...")
print(f"Number of customers: {final_df['ê³ ê°ì½”ë“œ'].nunique()}")
contract_df = rodb.execute_query(contract_query)

if contract_df is not None:
    print(f"âœ… Contract data fetched: {len(contract_df)} records")
    display(contract_df.head())
else:
    print("âŒ Failed to fetch contract data")
````

````python
# Step 2: ë°ì´í„° ì „ì²˜ë¦¬ ë° ë…¼ë¦¬ì  ê³„ì•½í•´ì§€ì¼ ê³„ì‚°
import numpy as np

# ë‚ ì§œ ì»¬ëŸ¼ ì „ì²˜ë¦¬
contract_df['ì‹œì‘ì¼ì'] = pd.to_datetime(contract_df['ì‹œì‘ì¼ì'], errors='coerce')
contract_df['ì¢…ë£Œì¼ì'] = pd.to_datetime(contract_df['ì¢…ë£Œì¼ì'], errors='coerce')
contract_df['í•´ì•½ì¼ì_SM'] = pd.to_datetime(contract_df['í•´ì•½ì¼ì_SM'], errors='coerce')

# ë…¼ë¦¬ì  ê³„ì•½í•´ì§€ì¼ ê³„ì‚° (í•´ì•½ì¼ìê°€ ìˆìœ¼ë©´ í•´ì•½ì¼ì, ì—†ìœ¼ë©´ ì¢…ë£Œì¼ì)
contract_df['ë…¼ë¦¬_ê³„ì•½í•´ì§€ì¼'] = contract_df['í•´ì•½ì¼ì_SM'].fillna(contract_df['ì¢…ë£Œì¼ì'])

# ìœ ì§€ì¼ìˆ˜ ê³„ì‚°
contract_df['ìœ ì§€ì¼ìˆ˜'] = (contract_df['ë…¼ë¦¬_ê³„ì•½í•´ì§€ì¼'] - contract_df['ì‹œì‘ì¼ì']).dt.days

print(f"Processed {len(contract_df)} contract records")
print(f"\ní•´ì•½ì¼ì_SMì´ ìˆëŠ” ê±´ìˆ˜: {contract_df['í•´ì•½ì¼ì_SM'].notna().sum()}")
print(f"í•´ì•½ì¼ì_SMì´ ì—†ëŠ” ê±´ìˆ˜: {contract_df['í•´ì•½ì¼ì_SM'].isna().sum()}")
contract_df[['ê³ ê°ì½”ë“œ', 'ê³„ì•½ì¼ë ¨ë²ˆí˜¸', 'ê³„ì•½ëŒ€ìƒ', 'ì‹œì‘ì¼ì', 'ì¢…ë£Œì¼ì', 'í•´ì•½ì¼ì_SM', 'ë…¼ë¦¬_ê³„ì•½í•´ì§€ì¼', 'ìœ ì§€ì¼ìˆ˜']].head(10)
````

````python
# Step 3: ê³„ì•½ëŒ€ìƒë³„ ì§‘ê³„ (ê³ ê°ì½”ë“œ + ê³„ì•½ëŒ€ìƒ)
# í•´ì•½ì—¬ë¶€ê´€ë ¨_ì •ë¦¬_v0_2.ipynbì˜ ë¡œì§ ì°¸ì¡°

contract_summary = contract_df.groupby(['ê³ ê°ì½”ë“œ','ê³„ì•½ëŒ€ìƒ']).agg({
    'ê³„ì•½ì¼ë ¨ë²ˆí˜¸': ['count','nunique'],
    'ì‹œì‘ì¼ì': 'min',
    'ì¢…ë£Œì¼ì': 'max',
    'í•´ì•½ì¼ì_SM': ['count','min','max'],
    'ë…¼ë¦¬_ê³„ì•½í•´ì§€ì¼': ['max','min'],
    'ìœ ì§€ì¼ìˆ˜': 'sum'
}).reset_index()

# ì»¬ëŸ¼ëª… ë‹¨ìˆœí™”
contract_summary.columns = ['ê³ ê°ì½”ë“œ', 'ê³„ì•½ëŒ€ìƒ', 'ê³„ì•½ì¼ë ¨ë²ˆí˜¸_cnt', 'ê³„ì•½ì¼ë ¨ë²ˆí˜¸_unicnt', 
                            'ìµœì´ˆê³„ì•½_ì‹œì‘ì¼ì', 'ìµœê·¼ê³„ì•½_ì¢…ë£Œì¼ì',
                            'í•´ì•½ì¼ì_SM_cnt', 'ìµœì´ˆ_í•´ì•½ì¼ì_SM', 'ìµœê·¼_í•´ì•½ì¼ì_SM', 
                            'ìµœê·¼_ë…¼ë¦¬_ê³„ì•½í•´ì§€ì¼', 'ìµœì´ˆ_ë…¼ë¦¬_ê³„ì•½í•´ì§€ì¼', 'ìœ ì§€ì¼ìˆ˜_sum']

# ê³„ì•½ëŒ€ìƒë³„ í•´ì•½ì—¬ë¶€ ê³„ì‚°: (ì´ ê³„ì•½ ê±´ìˆ˜ - í•´ì•½ ê±´ìˆ˜) > 0 ì´ë©´ 0(ìœ ì§€), ì•„ë‹ˆë©´ 1(í•´ì•½)
contract_summary['ê³„ì•½ëŒ€ìƒë³„_í•´ì•½ì—¬ë¶€'] = np.where(
    (contract_summary['ê³„ì•½ì¼ë ¨ë²ˆí˜¸_cnt'] - contract_summary['í•´ì•½ì¼ì_SM_cnt']) > 0, 
    0,  # ìœ ì§€ ì¤‘ì¸ ê³„ì•½ì´ ìˆìŒ
    1   # ëª¨ë“  ê³„ì•½ì´ í•´ì•½ë¨
)

print(f"ê³„ì•½ëŒ€ìƒë³„ ì§‘ê³„ ì™„ë£Œ: {len(contract_summary)} records")
print(f"\nê³„ì•½ëŒ€ìƒë³„_í•´ì•½ì—¬ë¶€ ë¶„í¬:")
print(contract_summary['ê³„ì•½ëŒ€ìƒë³„_í•´ì•½ì—¬ë¶€'].value_counts())
contract_summary.head(10)
````

````python
# Step 4: ê³„ì•½ëŒ€ìƒ ë¶„ë¥˜ ì •ë³´ merge
# ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜ëª… ì •ë³´ ì¶”ê°€

conttarg_info = contract_df[['ê³„ì•½ëŒ€ìƒ', 'Gerldesp', 'ê³„ì•½ëŒ€ìƒ_ì¤‘ë¶„ë¥˜ëª…', 'ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜ëª…']].drop_duplicates()
print(f"ê³„ì•½ëŒ€ìƒ ë¶„ë¥˜ ì •ë³´: {len(conttarg_info)} unique contract targets")

contract_summary_with_class = pd.merge(
    contract_summary, 
    conttarg_info, 
    on='ê³„ì•½ëŒ€ìƒ', 
    how='left'
)

# ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜ëª…ì´ ì—†ëŠ” ê±´ì€ ì œì™¸ (ì˜ˆì „ ê³„ì•½ê±´)
contract_summary_filtered = contract_summary_with_class[
    contract_summary_with_class['ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜ëª…'].notna()
].copy()

print(f"\ní•„í„°ë§ í›„: {len(contract_summary_filtered)} records")
print(f"ì œì™¸ëœ ê±´ìˆ˜: {len(contract_summary_with_class) - len(contract_summary_filtered)}")
print(f"\nê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜ëª… ë¶„í¬:")
print(contract_summary_filtered['ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜ëª…'].value_counts())
contract_summary_filtered.head()
````

````python
# Step 5: ê³„ì•½ëŒ€ìƒ ëŒ€ë¶„ë¥˜ë³„ ì§‘ê³„ (ê³ ê°ì½”ë“œ + ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜ëª…)
# ëŒ€ë¶„ë¥˜ë³„ë¡œ í•œ ê±´ì´ë¼ë„ ìœ ì§€ ì¤‘ì¸ ê³„ì•½ì´ ìˆìœ¼ë©´ 0(ìœ ì§€), ëª¨ë‘ í•´ì•½ì´ë©´ 1(í•´ì•½)

contract_by_category = contract_summary_filtered.groupby(['ê³ ê°ì½”ë“œ','ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜ëª…']).agg({
    'ê³„ì•½ëŒ€ìƒ': ['nunique', 'count'],
    'ìµœì´ˆê³„ì•½_ì‹œì‘ì¼ì': 'min',
    'ìµœê·¼ê³„ì•½_ì¢…ë£Œì¼ì': 'max',
    'ìµœê·¼_í•´ì•½ì¼ì_SM': 'max',
    'ìµœì´ˆ_í•´ì•½ì¼ì_SM': 'min',
    'í•´ì•½ì¼ì_SM_cnt': 'sum',
    'ê³„ì•½ì¼ë ¨ë²ˆí˜¸_unicnt': 'max',
    'ìµœê·¼_ë…¼ë¦¬_ê³„ì•½í•´ì§€ì¼': 'max',
    'ìµœì´ˆ_ë…¼ë¦¬_ê³„ì•½í•´ì§€ì¼': 'min',
    'ê³„ì•½ëŒ€ìƒë³„_í•´ì•½ì—¬ë¶€': 'sum',
}).reset_index()

# ì»¬ëŸ¼ëª… ë‹¨ìˆœí™”
contract_by_category.columns = ['ê³ ê°ì½”ë“œ', 'ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜ëª…', 'ê³„ì•½ëŒ€ìƒ_nunique', 'ê³„ì•½ëŒ€ìƒ_count',
                                'ìµœì´ˆê³„ì•½_ì‹œì‘ì¼ì', 'ìµœê·¼ê³„ì•½_ì¢…ë£Œì¼ì', 'ìµœê·¼_í•´ì•½ì¼ì_SM',
                                'ìµœì´ˆ_í•´ì•½ì¼ì_SM', 'í•´ì•½ì¼ì_SM_cnt', 'ê³„ì•½ì¼ë ¨ë²ˆí˜¸_unicnt',
                                'ìµœê·¼_ë…¼ë¦¬_ê³„ì•½í•´ì§€ì¼', 'ìµœì´ˆ_ë…¼ë¦¬_ê³„ì•½í•´ì§€ì¼', 'ê³„ì•½ëŒ€ìƒë³„_í•´ì•½ì—¬ë¶€_sum']

# ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜_í•´ì•½ì—¬ë¶€: (ê³„ì•½ëŒ€ìƒ ê°œìˆ˜ - í•´ì•½ëœ ê³„ì•½ëŒ€ìƒ ê°œìˆ˜) > 0 ì´ë©´ 0(ìœ ì§€), ì•„ë‹ˆë©´ 1(í•´ì•½)
contract_by_category['ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜_í•´ì•½ì—¬ë¶€'] = np.where(
    (contract_by_category['ê³„ì•½ëŒ€ìƒ_count'] - contract_by_category['ê³„ì•½ëŒ€ìƒë³„_í•´ì•½ì—¬ë¶€_sum']) > 0,
    0,  # ëŒ€ë¶„ë¥˜ ë‚´ ìœ ì§€ ì¤‘ì¸ ê³„ì•½ëŒ€ìƒì´ ìˆìŒ
    1   # ëŒ€ë¶„ë¥˜ ë‚´ ëª¨ë“  ê³„ì•½ëŒ€ìƒì´ í•´ì•½ë¨
)

print(f"ê³„ì•½ëŒ€ìƒ ëŒ€ë¶„ë¥˜ë³„ ì§‘ê³„: {len(contract_by_category)} records")
print(f"\nê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜_í•´ì•½ì—¬ë¶€ ë¶„í¬:")
print(contract_by_category['ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜_í•´ì•½ì—¬ë¶€'].value_counts())
contract_by_category.head(10)
````

````python
# Step 6: ìµœì¢… í•´ì•½ ì—¬ë¶€ ë¼ë²¨ ìƒì„± (ê³ ê°ì½”ë“œë³„)
# ëª¨ë“  ëŒ€ë¶„ë¥˜ê°€ í•´ì•½ë˜ë©´ 0(í•´ì•½), í•œ ê±´ì´ë¼ë„ ìœ ì§€ ì¤‘ì´ë©´ 1(ìœ ì§€)

final_churn_label = contract_by_category.groupby('ê³ ê°ì½”ë“œ').agg({
    'ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜ëª…': 'count',  # ëŒ€ë¶„ë¥˜ ê°œìˆ˜
    'ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜_í•´ì•½ì—¬ë¶€': 'sum'  # í•´ì•½ëœ ëŒ€ë¶„ë¥˜ ê°œìˆ˜
}).reset_index()

final_churn_label.columns = ['ê³ ê°ì½”ë“œ', 'ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜ëª…_count', 'ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜_í•´ì•½ì—¬ë¶€_sum']

# ìµœì¢…ê³„ì•½ìƒíƒœ: (ëŒ€ë¶„ë¥˜ ê°œìˆ˜ - í•´ì•½ëœ ëŒ€ë¶„ë¥˜ ê°œìˆ˜) > 0 ì´ë©´ 1(ìœ ì§€), ì•„ë‹ˆë©´ 0(í•´ì•½)
final_churn_label['ìµœì¢…ê³„ì•½ìƒíƒœ'] = np.where(
    (final_churn_label['ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜ëª…_count'] - final_churn_label['ê³„ì•½ëŒ€ìƒ_ëŒ€ë¶„ë¥˜_í•´ì•½ì—¬ë¶€_sum']) > 0,
    1,  # ìœ ì§€ (at least one active contract)
    0   # í•´ì•½ (all contracts cancelled)
)

# í•´ì•½ ì—¬ë¶€ ë¼ë²¨ ìƒì„± (1: í•´ì•½, 0: ìœ ì§€)
final_churn_label['í•´ì•½ì—¬ë¶€'] = 1 - final_churn_label['ìµœì¢…ê³„ì•½ìƒíƒœ']

print(f"ìµœì¢… í•´ì•½ ì—¬ë¶€ ë¼ë²¨ ìƒì„± ì™„ë£Œ: {len(final_churn_label)} customers")
print(f"\nìµœì¢…ê³„ì•½ìƒíƒœ ë¶„í¬:")
print(f"  1 (ìœ ì§€): {(final_churn_label['ìµœì¢…ê³„ì•½ìƒíƒœ'] == 1).sum()}")
print(f"  0 (í•´ì•½): {(final_churn_label['ìµœì¢…ê³„ì•½ìƒíƒœ'] == 0).sum()}")
print(f"\ní•´ì•½ì—¬ë¶€ ë¶„í¬:")
print(f"  1 (í•´ì•½): {(final_churn_label['í•´ì•½ì—¬ë¶€'] == 1).sum()}")
print(f"  0 (ìœ ì§€): {(final_churn_label['í•´ì•½ì—¬ë¶€'] == 0).sum()}")
print(f"\ní•´ì•½ìœ¨: {final_churn_label['í•´ì•½ì—¬ë¶€'].mean():.2%}")

final_churn_label.head(20)
````

````python
# Step 7: final_dfì— í•´ì•½ì—¬ë¶€ ë¼ë²¨ ì¶”ê°€
final_df_with_label = pd.merge(
    final_df,
    final_churn_label[['ê³ ê°ì½”ë“œ', 'í•´ì•½ì—¬ë¶€', 'ìµœì¢…ê³„ì•½ìƒíƒœ']],
    on='ê³ ê°ì½”ë“œ',
    how='left'
)

# ë¼ë²¨ì´ ì—†ëŠ” ê³ ê° í™•ì¸ (ê³„ì•½ ì •ë³´ê°€ ì—†ëŠ” ê³ ê°)
missing_label_count = final_df_with_label['í•´ì•½ì—¬ë¶€'].isna().sum()
print(f"ë¼ë²¨ ë§¤í•‘ ì™„ë£Œ:")
print(f"  ì „ì²´ ê³ ê° ìˆ˜: {len(final_df_with_label)}")
print(f"  ë¼ë²¨ì´ ìˆëŠ” ê³ ê° ìˆ˜: {(~final_df_with_label['í•´ì•½ì—¬ë¶€'].isna()).sum()}")
print(f"  ë¼ë²¨ì´ ì—†ëŠ” ê³ ê° ìˆ˜: {missing_label_count}")

if missing_label_count > 0:
    print(f"\nâš ï¸ ë¼ë²¨ì´ ì—†ëŠ” ê³ ê°ì€ ê³„ì•½ ì •ë³´ê°€ ì—†ëŠ” ê³ ê°ì…ë‹ˆë‹¤.")
    # ë¼ë²¨ì´ ì—†ëŠ” ê²½ìš° ê¸°ë³¸ê°’ 0(ìœ ì§€)ìœ¼ë¡œ ì„¤ì •
    final_df_with_label['í•´ì•½ì—¬ë¶€'] = final_df_with_label['í•´ì•½ì—¬ë¶€'].fillna(0).astype(int)
    final_df_with_label['ìµœì¢…ê³„ì•½ìƒíƒœ'] = final_df_with_label['ìµœì¢…ê³„ì•½ìƒíƒœ'].fillna(1).astype(int)

print(f"\nìµœì¢… ë°ì´í„°ì…‹:")
print(f"  Shape: {final_df_with_label.shape}")
print(f"  í•´ì•½ ê³ ê° ìˆ˜: {(final_df_with_label['í•´ì•½ì—¬ë¶€'] == 1).sum()}")
print(f"  ìœ ì§€ ê³ ê° ìˆ˜: {(final_df_with_label['í•´ì•½ì—¬ë¶€'] == 0).sum()}")
print(f"  í•´ì•½ìœ¨: {final_df_with_label['í•´ì•½ì—¬ë¶€'].mean():.2%}")

final_df_with_label[['ê³ ê°ì½”ë“œ', 'ê³ ê°ëª…', 'í•´ì•½ì—¬ë¶€', 'ìµœì¢…ê³„ì•½ìƒíƒœ']].head(20)
````

````python
# Step 3: Feature Engineering - ê³ ê° ìƒí˜¸ì‘ìš© íŠ¹ì„± (interaction_history)

print("\n" + "="*80)
print("ğŸ”§ Feature Engineering: ê³ ê° ìƒí˜¸ì‘ìš© íŠ¹ì„±")
print("="*80)

def extract_interaction_features(interaction_json):
    """ê³ ê° ìƒí˜¸ì‘ìš© ì´ë ¥ì—ì„œ íŠ¹ì„± ì¶”ì¶œ"""
    if pd.isna(interaction_json) or interaction_json == '':
        return {
            'interaction_count': 0,
            'interaction_complaint_count': 0,
            'interaction_inquiry_count': 0,
            'interaction_as_count': 0,
            'interaction_avg_response_days': 0,
            'interaction_max_response_days': 0,
            'interaction_days_since_last': np.nan,
            'interaction_has_unresolved': 0,
            'interaction_satisfaction_avg': 0,
        }
    
    try:
        if isinstance(interaction_json, str):
            interactions = json.loads(interaction_json)
        else:
            interactions = interaction_json
            
        if not isinstance(interactions, list):
            interactions = []
    except:
        interactions = []
    
    if len(interactions) == 0:
        return {
            'interaction_count': 0,
            'interaction_complaint_count': 0,
            'interaction_inquiry_count': 0,
            'interaction_as_count': 0,
            'interaction_avg_response_days': 0,
            'interaction_max_response_days': 0,
            'interaction_days_since_last': np.nan,
            'interaction_has_unresolved': 0,
            'interaction_satisfaction_avg': 0,
        }
    
    interaction_count = len(interactions)
    
    # ìœ í˜•ë³„ ì¹´ìš´íŠ¸ (í‚¤ê°€ ë‹¤ë¥¼ ìˆ˜ ìˆìœ¼ë¯€ë¡œ ì•ˆì „í•˜ê²Œ ì²˜ë¦¬)
    complaint_count = sum([1 for i in interactions if 'ë¶ˆë§Œ' in str(i.get('ìƒë‹´ìœ í˜•', '')) or 'í´ë ˆì„' in str(i.get('ìƒë‹´ìœ í˜•', ''))])
    inquiry_count = sum([1 for i in interactions if 'ë¬¸ì˜' in str(i.get('ìƒë‹´ìœ í˜•', ''))])
    as_count = sum([1 for i in interactions if 'AS' in str(i.get('ìƒë‹´ìœ í˜•', '')) or 'A/S' in str(i.get('ìƒë‹´ìœ í˜•', ''))])
    
    # ì‘ë‹µ ì‹œê°„ ê³„ì‚°
    response_times = []
    interaction_dates = []
    
    for i in interactions:
        try:
            # ì ‘ìˆ˜ì¼ê³¼ ì²˜ë¦¬ì¼ ì°¨ì´ ê³„ì‚°
            receipt_date = pd.to_datetime(i.get('ì ‘ìˆ˜ì¼ì') or i.get('ë“±ë¡ì¼ì'), errors='coerce')
            process_date = pd.to_datetime(i.get('ì²˜ë¦¬ì¼ì') or i.get('ì™„ë£Œì¼ì'), errors='coerce')
            
            if pd.notna(receipt_date):
                interaction_dates.append(receipt_date)
                
            if pd.notna(receipt_date) and pd.notna(process_date):
                response_time = (process_date - receipt_date).days
                if response_time >= 0:  # ìŒìˆ˜ ì œì™¸
                    response_times.append(response_time)
        except:
            pass
    
    avg_response_days = np.mean(response_times) if response_times else 0
    max_response_days = max(response_times) if response_times else 0
    
    # ìµœê·¼ ìƒí˜¸ì‘ìš©ìœ¼ë¡œë¶€í„° ê²½ê³¼ì¼ìˆ˜
    days_since_last = np.nan
    if interaction_dates:
        days_since_last = (pd.Timestamp.now() - max(interaction_dates)).days
    
    # ë¯¸í•´ê²° ê±´ ì—¬ë¶€
    has_unresolved = int(any([i.get('ì²˜ë¦¬ìƒíƒœ') == 'ë¯¸ì²˜ë¦¬' or i.get('ì™„ë£Œì—¬ë¶€') == 'N' for i in interactions]))
    
    # ë§Œì¡±ë„ í‰ê·  (ìˆëŠ” ê²½ìš°)
    satisfaction_scores = []
    for i in interactions:
        try:
            score = i.get('ë§Œì¡±ë„') or i.get('í‰ì ')
            if score and str(score).replace('.', '').isdigit():
                satisfaction_scores.append(float(score))
        except:
            pass
    
    satisfaction_avg = np.mean(satisfaction_scores) if satisfaction_scores else 0
    
    return {
        'interaction_count': interaction_count,
        'interaction_complaint_count': complaint_count,
        'interaction_inquiry_count': inquiry_count,
        'interaction_as_count': as_count,
        'interaction_avg_response_days': avg_response_days,
        'interaction_max_response_days': max_response_days,
        'interaction_days_since_last': days_since_last,
        'interaction_has_unresolved': has_unresolved,
        'interaction_satisfaction_avg': satisfaction_avg,
    }

# ìƒí˜¸ì‘ìš© íŠ¹ì„± ì¶”ì¶œ
print("Extracting interaction features...")
interaction_features = df_features['interaction_history'].apply(extract_interaction_features)
interaction_features_df = pd.DataFrame(list(interaction_features))

# ë°ì´í„°í”„ë ˆì„ì— ì¶”ê°€
for col in interaction_features_df.columns:
    df_features[col] = interaction_features_df[col]

print(f"âœ… Created {len(interaction_features_df.columns)} interaction features")
print(f"\nInteraction features:")
for col in interaction_features_df.columns:
    print(f"   - {col}")

print(f"\nê¸°ë³¸ í†µê³„:")
print(interaction_features_df.describe().T[['mean', 'std', 'min', 'max']])
````

````python
# Step 4: Feature Engineering - ì‘ì—… ì´ë ¥ íŠ¹ì„± (ì‘ì—…ì´ë ¥)

print("\n" + "="*80)
print("ğŸ”§ Feature Engineering: ì‘ì—… ì´ë ¥ íŠ¹ì„±")
print("="*80)

def extract_work_features(work_json):
    """ì‘ì—… ì´ë ¥ì—ì„œ íŠ¹ì„± ì¶”ì¶œ"""
    if pd.isna(work_json) or work_json == '':
        return {
            'work_count': 0,
            'work_confirmed_count': 0,
            'work_cancelled_count': 0,
            'work_unique_types': 0,
            'work_days_since_last': np.nan,
            'work_days_since_first': np.nan,
            'work_frequency_days': 0,  # í‰ê·  ì‘ì—… ë¹ˆë„ (ì¼)
            'work_avg_csi_score': 0,
            'work_has_survey': 0,
            'work_pest_detected_count': 0,
            'work_avg_work_duration_min': 0,
            'work_total_divisions': 0,  # ì´ êµ¬íš ìˆ˜
        }
    
    try:
        if isinstance(work_json, str):
            works = json.loads(work_json)
        else:
            works = work_json
            
        if not isinstance(works, list):
            works = []
    except:
        works = []
    
    if len(works) == 0:
        return {
            'work_count': 0,
            'work_confirmed_count': 0,
            'work_cancelled_count': 0,
            'work_unique_types': 0,
            'work_days_since_last': np.nan,
            'work_days_since_first': np.nan,
            'work_frequency_days': 0,
            'work_avg_csi_score': 0,
            'work_has_survey': 0,
            'work_pest_detected_count': 0,
            'work_avg_work_duration_min': 0,
            'work_total_divisions': 0,
        }
    
    work_count = len(works)
    
    # í™•ì •/ì·¨ì†Œ ì‘ì—… ìˆ˜
    confirmed_count = sum([1 for w in works if w.get('í™•ì •ì—¬ë¶€') == 'Y'])
    cancelled_count = sum([1 for w in works if w.get('ì·¨ì†Œì—¬ë¶€') == 'Y'])
    
    # ê³ ìœ  ì‘ì—… ìœ í˜• ìˆ˜
    unique_types = len(set([w.get('ì‘ì—…ìœ í˜•') for w in works if w.get('ì‘ì—…ìœ í˜•')]))
    
    # ì‘ì—… ì¼ì ë¶„ì„
    work_dates = []
    for w in works:
        try:
            work_date = pd.to_datetime(w.get('ì‘ì—…ì¼ì'), errors='coerce')
            if pd.notna(work_date):
                work_dates.append(work_date)
        except:
            pass
    
    days_since_last = np.nan
    days_since_first = np.nan
    work_frequency_days = 0
    
    if work_dates:
        work_dates_sorted = sorted(work_dates)
        days_since_last = (pd.Timestamp.now() - work_dates_sorted[-1]).days
        days_since_first = (pd.Timestamp.now() - work_dates_sorted[0]).days
        
        # í‰ê·  ì‘ì—… ë¹ˆë„ ê³„ì‚°
        if len(work_dates_sorted) > 1:
            total_span = (work_dates_sorted[-1] - work_dates_sorted[0]).days
            work_frequency_days = total_span / (len(work_dates_sorted) - 1)
    
    # CSI ì ìˆ˜ (ì„œë¹„ìŠ¤ ë§Œì¡±ë„)
    csi_scores = []
    has_survey = 0
    
    for w in works:
        try:
            survey_data = w.get('ì„œë¹„ìŠ¤_ë§Œì¡±ë„')
            if survey_data:
                has_survey = 1
                if isinstance(survey_data, str):
                    survey_list = json.loads(survey_data)
                else:
                    survey_list = survey_data
                    
                if isinstance(survey_list, list):
                    for s in survey_list:
                        score = s.get('í‰ê· _CSI_ì ìˆ˜')
                        if score and str(score).replace('.', '').replace('-', '').isdigit():
                            csi_scores.append(float(score))
        except:
            pass
    
    avg_csi_score = np.mean(csi_scores) if csi_scores else 0
    
    # í•´ì¶© íƒì§€ ì •ë³´
    pest_detected_count = 0
    work_durations = []
    total_divisions = 0
    
    for w in works:
        try:
            # ì‘ì—… ë‚´ì—­ì—ì„œ ê°œì¸ ì‘ì—… ë¶„ì„
            work_details = w.get('ì‘ì—…ë‚´ì—­')
            if work_details:
                if isinstance(work_details, str):
                    work_details_list = json.loads(work_details)
                else:
                    work_details_list = work_details
                
                if isinstance(work_details_list, list):
                    for wd in work_details_list:
                        # ì‘ì—… ì‹œê°„ ê³„ì‚°
                        try:
                            fr_time = wd.get('ì„œë¹„ìŠ¤ì‹œì‘ì‹œê°„')
                            to_time = wd.get('ì„œë¹„ìŠ¤ì¢…ë£Œì‹œê°„')
                            if fr_time and to_time:
                                fr = pd.to_datetime(fr_time, format='%H%M', errors='coerce')
                                to = pd.to_datetime(to_time, format='%H%M', errors='coerce')
                                if pd.notna(fr) and pd.notna(to):
                                    duration = (to - fr).total_seconds() / 60
                                    if duration > 0:
                                        work_durations.append(duration)
                        except:
                            pass
                        
                        # ê°œì¸ì‘ì—…ìƒì„¸ì—ì„œ í•´ì¶© ì •ë³´
                        work_detail = wd.get('ê°œì¸ì‘ì—…ìƒì„¸')
                        if work_detail:
                            if isinstance(work_detail, str):
                                detail_list = json.loads(work_detail)
                            else:
                                detail_list = work_detail
                            
                            if isinstance(detail_list, list):
                                for d in detail_list:
                                    # í•´ì¶©ì´ ë°œê²¬ëœ ê²½ìš°
                                    pest_cnt = d.get('ê°œì²´ìˆ˜') or d.get('í•´ì¶©ë§ˆë¦¬ìˆ˜') or 0
                                    if pest_cnt and int(pest_cnt) > 0:
                                        pest_detected_count += 1
                                    
                                    # êµ¬íš ìˆ˜
                                    if d.get('êµ¬íšë²ˆí˜¸'):
                                        total_divisions += 1
        except:
            pass
    
    avg_work_duration = np.mean(work_durations) if work_durations else 0
    
    return {
        'work_count': work_count,
        'work_confirmed_count': confirmed_count,
        'work_cancelled_count': cancelled_count,
        'work_unique_types': unique_types,
        'work_days_since_last': days_since_last,
        'work_days_since_first': days_since_first,
        'work_frequency_days': work_frequency_days,
        'work_avg_csi_score': avg_csi_score,
        'work_has_survey': has_survey,
        'work_pest_detected_count': pest_detected_count,
        'work_avg_work_duration_min': avg_work_duration,
        'work_total_divisions': total_divisions,
    }

# ì‘ì—… ì´ë ¥ íŠ¹ì„± ì¶”ì¶œ
print("Extracting work history features...")
work_features = df_features['ì‘ì—…ì´ë ¥'].apply(extract_work_features)
work_features_df = pd.DataFrame(list(work_features))

# ë°ì´í„°í”„ë ˆì„ì— ì¶”ê°€
for col in work_features_df.columns:
    df_features[col] = work_features_df[col]

print(f"âœ… Created {len(work_features_df.columns)} work history features")
print(f"\nWork history features:")
for col in work_features_df.columns:
    print(f"   - {col}")

print(f"\nê¸°ë³¸ í†µê³„:")
print(work_features_df.describe().T[['mean', 'std', 'min', 'max']])
````

````python
# Step 5: Feature Engineering - ê¸°ë³¸ ê³ ê° ì •ë³´ ë° íŒŒìƒ íŠ¹ì„±

print("\n" + "="*80)
print("ğŸ”§ Feature Engineering: ê¸°ë³¸ ì •ë³´ ë° íŒŒìƒ íŠ¹ì„±")
print("="*80)

# 1. ê³ ê° ë“±ë¡ì¼ë¡œë¶€í„° ê²½ê³¼ì¼ìˆ˜ (ê³ ê° ì—°ë ¹)
df_features['customer_age_days'] = (pd.Timestamp.now() - pd.to_datetime(df_features['ë“±ë¡ì¼ì'], errors='coerce')).dt.days

# 2. ì§€ë¦¬ì  íŠ¹ì„±
df_features['has_geo_info'] = ((df_features['ìœ„ë„'].notna()) & (df_features['ê²½ë„'].notna())).astype(int)
df_features['has_utm_info'] = ((df_features['UTMK_X'].notna()) & (df_features['UTMK_Y'].notna())).astype(int)

# 3. ì£¼ì†Œ ì •ë³´
df_features['has_postal_code'] = df_features['ìš°í¸ë²ˆí˜¸'].notna().astype(int)
df_features['address_length'] = df_features['ì£¼ì†Œ1'].fillna('').str.len()

# 4. ì‚¬ì—…ì ì •ë³´
df_features['has_business_no'] = df_features['ì‚¬ì—…ìë²ˆí˜¸'].notna().astype(int)
df_features['ì—…íƒœ_length'] = df_features['ì—…íƒœ'].fillna('').str.len()
df_features['ì¢…ëª©_length'] = df_features['ì¢…ëª©'].fillna('').str.len()

# 5. ë¯¸ìˆ˜ ë‚´ì—­
df_features['has_ë¯¸ìˆ˜'] = df_features['ë¯¸ìˆ˜ë‚´ì—­'].notna().astype(int)

# 6. ê³„ì•½ ê´€ë ¨ íŒŒìƒ íŠ¹ì„±
df_features['contract_active_ratio'] = np.where(
    df_features['contract_count'] > 0,
    df_features['contract_active_count'] / df_features['contract_count'],
    0
)

df_features['contract_cancelled_ratio'] = np.where(
    df_features['contract_count'] > 0,
    df_features['contract_cancelled_count'] / df_features['contract_count'],
    0
)

# 7. ì‘ì—… ë¹ˆë„ ê´€ë ¨ íŠ¹ì„±
df_features['work_per_contract'] = np.where(
    df_features['contract_count'] > 0,
    df_features['work_count'] / df_features['contract_count'],
    0
)

# 8. ìµœê·¼ì„± ì§€í‘œ (Recency)
df_features['recency_work_vs_contract'] = df_features['work_days_since_last'] - df_features['contract_days_since_last']

# 9. ê³ ê° í™œë™ì„± ì ìˆ˜
df_features['customer_activity_score'] = (
    (df_features['contract_count'] > 0).astype(int) +
    (df_features['work_count'] > 0).astype(int) +
    (df_features['interaction_count'] > 0).astype(int) +
    (df_features['work_has_survey'] == 1).astype(int)
)

# 10. ê³„ì•½ ë‹¤ì–‘ì„± ì§€í‘œ
df_features['contract_diversity'] = (
    df_features['contract_ëŒ€ë¶„ë¥˜_count'] * 2 + 
    df_features['contract_ì¤‘ë¶„ë¥˜_count']
)

print(f"âœ… Created additional derived features")

# ìƒˆë¡œ ìƒì„±ëœ íŠ¹ì„± ë¦¬ìŠ¤íŠ¸
new_features = [
    'customer_age_days', 'has_geo_info', 'has_utm_info', 'has_postal_code',
    'address_length', 'has_business_no', 'ì—…íƒœ_length', 'ì¢…ëª©_length',
    'has_ë¯¸ìˆ˜', 'contract_active_ratio', 'contract_cancelled_ratio',
    'work_per_contract', 'recency_work_vs_contract', 'customer_activity_score',
    'contract_diversity'
]

print(f"\nDerived features:")
for feat in new_features:
    print(f"   - {feat}")

print(f"\nê¸°ë³¸ í†µê³„:")
print(df_features[new_features].describe().T[['mean', 'std', 'min', 'max']])
````

````python
# Step 6: Categorical Features - Label Encoding

print("\n" + "="*80)
print("ğŸ”§ Feature Engineering: ë²”ì£¼í˜• ë³€ìˆ˜ ì¸ì½”ë”©")
print("="*80)

from sklearn.preprocessing import LabelEncoder

# ì¸ì½”ë”©í•  ë²”ì£¼í˜• ë³€ìˆ˜
categorical_features = {
    'ìœ í˜•ëŒ€': 'customer_type_large_encoded',
    'ìœ í˜•ì¤‘': 'customer_type_medium_encoded',
    'ìœ í˜•ëŒ€_ëª…ì¹­': 'customer_type_large_name_encoded',
    'ìœ í˜•ì¤‘_ëª…ì¹­': 'customer_type_medium_name_encoded',
    'ëŒ€ë¶„ë¥˜': 'industry_large_encoded',
    'ì¤‘ë¶„ë¥˜': 'industry_medium_encoded',
    'ì†Œë¶„ë¥˜': 'industry_small_encoded',
    'ì„±ë³„': 'gender_encoded',
    'êµ­ì ': 'nationality_encoded',
    'ì‹ ê³ ê°ë¶„ë¥˜ì½”ë“œ': 'new_customer_class_encoded',
}

label_encoders = {}

for original_col, encoded_col in categorical_features.items():
    if original_col in df_features.columns:
        # ê²°ì¸¡ê°’ì„ 'UNKNOWN'ìœ¼ë¡œ ì²˜ë¦¬
        df_features[f'{original_col}_filled'] = df_features[original_col].fillna('UNKNOWN').astype(str)
        
        # Label Encoding
        le = LabelEncoder()
        df_features[encoded_col] = le.fit_transform(df_features[f'{original_col}_filled'])
        label_encoders[original_col] = le
        
        # ì„ì‹œ ì»¬ëŸ¼ ì œê±°
        df_features.drop(f'{original_col}_filled', axis=1, inplace=True)
        
        print(f"   - {original_col} â†’ {encoded_col} ({len(le.classes_)} classes)")

print(f"\nâœ… Encoded {len(categorical_features)} categorical features")

# ë²”ì£¼í˜• ë³€ìˆ˜ì˜ ê³ ìœ ê°’ ìˆ˜
print(f"\në²”ì£¼í˜• ë³€ìˆ˜ ê³ ìœ ê°’ ìˆ˜:")
for original_col, encoded_col in categorical_features.items():
    if encoded_col in df_features.columns:
        n_unique = df_features[encoded_col].nunique()
        print(f"   - {encoded_col}: {n_unique} unique values")
````

````python
# Step 7: Feature Summary ë° ìµœì¢… Feature Set ìƒì„±

print("\n" + "="*80)
print("ğŸ“Š Feature Engineering ì™„ë£Œ - ì „ì²´ íŠ¹ì„± ìš”ì•½")
print("="*80)

# ëª¨ë“  ìƒì„±ëœ íŠ¹ì„± ë¦¬ìŠ¤íŠ¸
all_created_features = []

# 1. Contract features
contract_feat = [
    'contract_count', 'contract_total_area', 'contract_total_area_m',
    'contract_avg_duration_days', 'contract_max_duration_days', 'contract_min_duration_days',
    'contract_active_count', 'contract_cancelled_count', 'contract_unique_targets',
    'contract_has_special_note', 'contract_days_since_first', 'contract_days_since_last',
    'contract_ëŒ€ë¶„ë¥˜_count', 'contract_ì¤‘ë¶„ë¥˜_count'
]
all_created_features.extend(contract_feat)

# 2. Interaction features
interaction_feat = [
    'interaction_count', 'interaction_complaint_count', 'interaction_inquiry_count',
    'interaction_as_count', 'interaction_avg_response_days', 'interaction_max_response_days',
    'interaction_days_since_last', 'interaction_has_unresolved', 'interaction_satisfaction_avg'
]
all_created_features.extend(interaction_feat)

# 3. Work history features
work_feat = [
    'work_count', 'work_confirmed_count', 'work_cancelled_count', 'work_unique_types',
    'work_days_since_last', 'work_days_since_first', 'work_frequency_days',
    'work_avg_csi_score', 'work_has_survey', 'work_pest_detected_count',
    'work_avg_work_duration_min', 'work_total_divisions'
]
all_created_features.extend(work_feat)

# 4. Derived features
derived_feat = [
    'customer_age_days', 'has_geo_info', 'has_utm_info', 'has_postal_code',
    'address_length', 'has_business_no', 'ì—…íƒœ_length', 'ì¢…ëª©_length',
    'has_ë¯¸ìˆ˜', 'contract_active_ratio', 'contract_cancelled_ratio',
    'work_per_contract', 'recency_work_vs_contract', 'customer_activity_score',
    'contract_diversity'
]
all_created_features.extend(derived_feat)

# 5. Encoded categorical features
encoded_feat = [
    'customer_type_large_encoded', 'customer_type_medium_encoded',
    'customer_type_large_name_encoded', 'customer_type_medium_name_encoded',
    'industry_large_encoded', 'industry_medium_encoded', 'industry_small_encoded',
    'gender_encoded', 'nationality_encoded', 'new_customer_class_encoded'
]
all_created_features.extend(encoded_feat)

# 6. Geographic features (existing)
geo_feat = ['ìœ„ë„', 'ê²½ë„', 'UTMK_X', 'UTMK_Y']
all_created_features.extend(geo_feat)

# 7. Target variable
target_feat = ['í•´ì•½ì—¬ë¶€']

print(f"\nğŸ“‹ ìƒì„±ëœ íŠ¹ì„± ì¹´í…Œê³ ë¦¬ë³„ ìš”ì•½:")
print(f"   1. ê³„ì•½ ì •ë³´ íŠ¹ì„± (Contract):        {len(contract_feat)}ê°œ")
print(f"   2. ìƒí˜¸ì‘ìš© íŠ¹ì„± (Interaction):      {len(interaction_feat)}ê°œ")
print(f"   3. ì‘ì—… ì´ë ¥ íŠ¹ì„± (Work History):    {len(work_feat)}ê°œ")
print(f"   4. íŒŒìƒ íŠ¹ì„± (Derived):             {len(derived_feat)}ê°œ")
print(f"   5. ì¸ì½”ë”©ëœ ë²”ì£¼í˜• (Encoded):        {len(encoded_feat)}ê°œ")
print(f"   6. ì§€ë¦¬ì  íŠ¹ì„± (Geographic):         {len(geo_feat)}ê°œ")
print(f"   {'â”€'*60}")
print(f"   ì´ íŠ¹ì„± ìˆ˜ (Features):              {len(all_created_features)}ê°œ")
print(f"   íƒ€ê²Ÿ ë³€ìˆ˜ (Target):                  {len(target_feat)}ê°œ")

# ë¨¸ì‹ ëŸ¬ë‹ìš© ìµœì¢… ë°ì´í„°ì…‹ ìƒì„±
ml_features = all_created_features.copy()
ml_dataset = df_features[ml_features + target_feat + ['ê³ ê°ì½”ë“œ', 'ê³ ê°ëª…']].copy()

print(f"\nğŸ“Š ìµœì¢… ML ë°ì´í„°ì…‹:")
print(f"   - Shape: {ml_dataset.shape}")
print(f"   - Features: {len(ml_features)}")
print(f"   - Samples: {len(ml_dataset)}")
print(f"   - Target: í•´ì•½ì—¬ë¶€")
print(f"   - Churn rate: {ml_dataset['í•´ì•½ì—¬ë¶€'].mean():.2%}")

# ê²°ì¸¡ê°’ í™•ì¸
missing_counts = ml_dataset[ml_features].isnull().sum()
features_with_missing = missing_counts[missing_counts > 0].sort_values(ascending=False)

print(f"\nâš ï¸ ê²°ì¸¡ê°’ì´ ìˆëŠ” íŠ¹ì„± ({len(features_with_missing)}ê°œ):")
if len(features_with_missing) > 0:
    for feat, count in features_with_missing.head(10).items():
        print(f"   - {feat}: {count} ({count/len(ml_dataset)*100:.1f}%)")
else:
    print("   ì—†ìŒ")

# ë°ì´í„° íƒ€ì… í™•ì¸
print(f"\nğŸ“Œ ë°ì´í„° íƒ€ì… ë¶„í¬:")
dtype_counts = ml_dataset[ml_features].dtypes.value_counts()
for dtype, count in dtype_counts.items():
    print(f"   - {dtype}: {count}ê°œ")

print("\n" + "="*80)
````

````python
# Fix interaction_days_since_last by using the correct date field name

def extract_interaction_features_fixed(interaction_json):
    """Extract features from interaction_history JSON with corrected date field"""
    from datetime import datetime
    
    features = {}
    
    if pd.isna(interaction_json) or interaction_json == '' or interaction_json == '[]':
        features['interaction_count'] = 0
        features['interaction_voc_count'] = 0
        features['interaction_ìƒë‹´_count'] = 0
        features['interaction_ë¯¼ì›_count'] = 0
        features['interaction_type_unique'] = 0
        features['interaction_avg_length'] = 0
        features['interaction_has_ì ‘ìˆ˜ë‚´ìš©'] = 0
        features['interaction_recent_year'] = None
        features['interaction_days_since_last'] = None  # Will be filled
        return features
    
    try:
        # Parse JSON
        if isinstance(interaction_json, str):
            interactions = json.loads(interaction_json)
        else:
            interactions = interaction_json
            
        if not isinstance(interactions, list) or len(interactions) == 0:
            features['interaction_count'] = 0
            features['interaction_voc_count'] = 0
            features['interaction_ìƒë‹´_count'] = 0
            features['interaction_ë¯¼ì›_count'] = 0
            features['interaction_type_unique'] = 0
            features['interaction_avg_length'] = 0
            features['interaction_has_ì ‘ìˆ˜ë‚´ìš©'] = 0
            features['interaction_recent_year'] = None
            features['interaction_days_since_last'] = None
            return features
        
        # Count features
        features['interaction_count'] = len(interactions)
        
        # Event type counts
        event_types = [item.get('event_type', '') for item in interactions]
        features['interaction_voc_count'] = event_types.count('VOC')
        features['interaction_ìƒë‹´_count'] = event_types.count('ìƒë‹´')
        features['interaction_ë¯¼ì›_count'] = event_types.count('ë¯¼ì›')
        features['interaction_type_unique'] = len(set(event_types))
        
        # Content features
        contents = [item.get('ì ‘ìˆ˜ë‚´ìš©', '') for item in interactions]
        valid_contents = [c for c in contents if c and c.strip()]
        features['interaction_avg_length'] = np.mean([len(str(c)) for c in valid_contents]) if valid_contents else 0
        features['interaction_has_ì ‘ìˆ˜ë‚´ìš©'] = 1 if valid_contents else 0
        
        # Date features - using 'ì ‘ìˆ˜ì¼ì‹œ' field
        dates = []
        for item in interactions:
            date_str = item.get('ì ‘ìˆ˜ì¼ì‹œ', '')  # Korean field name!
            if date_str:
                try:
                    # Parse ISO format: 2021-12-10T00:00:00
                    if 'T' in date_str:
                        date_obj = datetime.fromisoformat(date_str)
                    else:
                        date_obj = datetime.strptime(date_str, '%Y-%m-%d')
                    dates.append(date_obj)
                except:
                    pass
        
        if dates:
            most_recent_date = max(dates)
            features['interaction_recent_year'] = most_recent_date.year
            days_since = (datetime.now() - most_recent_date).days
            features['interaction_days_since_last'] = days_since
        else:
            features['interaction_recent_year'] = None
            features['interaction_days_since_last'] = None
            
    except Exception as e:
        # Return default values on error
        features['interaction_count'] = 0
        features['interaction_voc_count'] = 0
        features['interaction_ìƒë‹´_count'] = 0
        features['interaction_ë¯¼ì›_count'] = 0
        features['interaction_type_unique'] = 0
        features['interaction_avg_length'] = 0
        features['interaction_has_ì ‘ìˆ˜ë‚´ìš©'] = 0
        features['interaction_recent_year'] = None
        features['interaction_days_since_last'] = None
    
    return features

print("ğŸ”§ Fixed interaction feature extraction function")
print("   - Now correctly parsing 'ì ‘ìˆ˜ì¼ì‹œ' field (Korean field name)")
print("   - This should fill interaction_days_since_last properly")
````

````python
# Create additional derived features for churn prediction

print("="*80)
print("ğŸš€ Creating Additional Derived Features")
print("="*80)

original_feature_count = len(ml_dataset.columns)

# ============================================================================
# 1. RATIO & COMPARISON FEATURES
# ============================================================================
print("\n1ï¸âƒ£ Ratio & Comparison Features")

# Interaction to work ratio (engagement balance)
ml_dataset['interaction_to_work_ratio'] = ml_dataset['interaction_count'] / (ml_dataset['work_count'] + 1)
print("   âœ“ interaction_to_work_ratio: Balance between complaints and service visits")

# VOC intensity (complaint rate)
ml_dataset['voc_intensity'] = ml_dataset['interaction_voc_count'] / (ml_dataset['interaction_count'] + 1)
print("   âœ“ voc_intensity: Proportion of VOC among all interactions")

# Work cancellation rate (service unreliability indicator)
ml_dataset['work_cancellation_rate'] = ml_dataset['work_cancelled_count'] / (ml_dataset['work_count'] + 1)
print("   âœ“ work_cancellation_rate: Proportion of cancelled service visits")

# Contract cancellation rate
ml_dataset['contract_cancellation_rate'] = ml_dataset['contract_cancelled_count'] / (ml_dataset['contract_count'] + 1)
print("   âœ“ contract_cancellation_rate: Historical cancellation tendency")

# Active contract ratio
ml_dataset['active_contract_ratio'] = ml_dataset['contract_active_count'] / (ml_dataset['contract_count'] + 1)
print("   âœ“ active_contract_ratio: Proportion of currently active contracts")


# ============================================================================
# 2. LIFECYCLE & MATURITY FEATURES
# ============================================================================
print("\n2ï¸âƒ£ Lifecycle & Maturity Features")

# Customer age in years
ml_dataset['customer_age_years'] = ml_dataset['contract_days_since_first'] / 365.25
print("   âœ“ customer_age_years: How long customer has been with company")

# Service history length in years
ml_dataset['service_history_years'] = ml_dataset['work_days_since_first'] / 365.25
print("   âœ“ service_history_years: Length of service relationship")

# Recent activity window (last 90 days indicator)
ml_dataset['recently_active_interaction'] = (ml_dataset['interaction_days_since_last'] <= 90).astype(int)
ml_dataset['recently_active_work'] = (ml_dataset['work_days_since_last'] <= 90).astype(int)
print("   âœ“ recently_active_* : Binary indicators for activity in last 90 days")

# Dormant customer indicators
ml_dataset['is_dormant_interaction'] = (ml_dataset['interaction_days_since_last'] > 365).astype(int)
ml_dataset['is_dormant_work'] = (ml_dataset['work_days_since_last'] > 365).astype(int)
print("   âœ“ is_dormant_* : Indicators for dormant customers (>1 year)")


# ============================================================================
# 3. ENGAGEMENT INTENSITY FEATURES
# ============================================================================
print("\n3ï¸âƒ£ Engagement Intensity Features")

# Average interactions per year
ml_dataset['interaction_per_year'] = ml_dataset['interaction_count'] / (ml_dataset['customer_age_years'] + 0.1)
print("   âœ“ interaction_per_year: Interaction frequency normalized by customer age")

# Average works per year
ml_dataset['work_per_year'] = ml_dataset['work_count'] / (ml_dataset['customer_age_years'] + 0.1)
print("   âœ“ work_per_year: Service visit frequency normalized by customer age")

# Total engagement score (weighted combination)
ml_dataset['total_engagement_score'] = (
    ml_dataset['interaction_count'] * 0.3 +  # Interactions weighted less
    ml_dataset['work_count'] * 0.7  # Service visits weighted more
)
print("   âœ“ total_engagement_score: Weighted combination of all engagements")

# Engagement diversity (unique types)
ml_dataset['engagement_diversity'] = (
    ml_dataset['interaction_type_unique'] + 
    ml_dataset['work_unique_types'] +
    ml_dataset['contract_unique_targets']
)
print("   âœ“ engagement_diversity: Variety of engagement types")


# ============================================================================
# 4. SERVICE QUALITY INDICATORS
# ============================================================================
print("\n4ï¸âƒ£ Service Quality Indicators")

# Work completion rate (already exists but let's ensure it)
if 'work_confirmed_count' in ml_dataset.columns:
    ml_dataset['work_completion_rate'] = ml_dataset['work_confirmed_count'] / (ml_dataset['work_count'] + 1)
    print("   âœ“ work_completion_rate: Proportion of completed works")

# CSI reliability (has survey data)
if 'work_has_survey' in ml_dataset.columns and 'work_avg_csi_score' in ml_dataset.columns:
    ml_dataset['csi_reliability'] = ml_dataset['work_has_survey'].astype(int)
    ml_dataset['csi_quality'] = ml_dataset['work_avg_csi_score'].fillna(0)
    print("   âœ“ csi_reliability, csi_quality: Customer satisfaction indicators")

# Pest detection rate (service effectiveness)
if 'work_pest_detected_count' in ml_dataset.columns:
    ml_dataset['pest_detection_rate'] = ml_dataset['work_pest_detected_count'] / (ml_dataset['work_count'] + 1)
    print("   âœ“ pest_detection_rate: Proportion of visits that found pests")


# ============================================================================
# 5. RISK INDICATORS
# ============================================================================
print("\n5ï¸âƒ£ Risk Indicators")

# Churn risk score based on recency (longer since last contact = higher risk)
ml_dataset['risk_score_recency'] = (
    ml_dataset['interaction_days_since_last'].fillna(365) / 365 * 0.4 +
    ml_dataset['work_days_since_last'].fillna(365) / 365 * 0.6
)
print("   âœ“ risk_score_recency: Combined recency risk (0=active, higher=risky)")

# Churn risk score based on cancellations
ml_dataset['risk_score_cancellation'] = (
    ml_dataset['contract_cancellation_rate'] * 0.6 +
    ml_dataset['work_cancellation_rate'] * 0.4
)
print("   âœ“ risk_score_cancellation: Historical cancellation tendency")

# Engagement decline indicator (comparing recent vs historical)
ml_dataset['engagement_decline'] = (
    (ml_dataset['interaction_days_since_last'] > ml_dataset['interaction_days_since_last'].median()).astype(int) +
    (ml_dataset['work_days_since_last'] > ml_dataset['work_days_since_last'].median()).astype(int)
) / 2
print("   âœ“ engagement_decline: Indicator if customer is less active than average")

# Combined risk score
ml_dataset['combined_risk_score'] = (
    ml_dataset['risk_score_recency'] * 0.4 +
    ml_dataset['risk_score_cancellation'] * 0.3 +
    ml_dataset['engagement_decline'] * 0.3
)
print("   âœ“ combined_risk_score: Comprehensive risk indicator")


# ============================================================================
# 6. TEMPORAL PATTERN FEATURES
# ============================================================================
print("\n6ï¸âƒ£ Temporal Pattern Features")

# Recent year of activity
if 'interaction_recent_year' in ml_dataset.columns:
    ml_dataset['years_since_last_interaction'] = 2025 - ml_dataset['interaction_recent_year'].fillna(2000)
    print("   âœ“ years_since_last_interaction: Years since most recent interaction")

# Service frequency (visits per unit time)
ml_dataset['service_frequency'] = ml_dataset['work_count'] / (ml_dataset['work_days_since_first'] / 30 + 1)
print("   âœ“ service_frequency: Average works per month of service history")

# Interaction frequency
ml_dataset['interaction_frequency'] = ml_dataset['interaction_count'] / (ml_dataset['customer_age_years'] * 12 + 1)
print("   âœ“ interaction_frequency: Average interactions per month")

# Time between services (average)
if 'work_frequency_days' in ml_dataset.columns:
    ml_dataset['avg_days_between_services'] = ml_dataset['work_frequency_days']
    print("   âœ“ avg_days_between_services: Average gap between service visits")


# ============================================================================
# 7. CONTRACT-SPECIFIC FEATURES
# ============================================================================
print("\n7ï¸âƒ£ Contract-Specific Features")

# Contract portfolio size (total managed area)
ml_dataset['portfolio_area_total'] = ml_dataset['contract_total_area'].fillna(0)
ml_dataset['portfolio_area_m_total'] = ml_dataset['contract_total_area_m'].fillna(0)
print("   âœ“ portfolio_area_*: Total area under contract")

# Average contract area
ml_dataset['avg_contract_area'] = ml_dataset['contract_total_area'] / (ml_dataset['contract_count'] + 1)
ml_dataset['avg_contract_area_m'] = ml_dataset['contract_total_area_m'] / (ml_dataset['contract_count'] + 1)
print("   âœ“ avg_contract_area_*: Average size per contract")

# Contract complexity (duration variance)
if 'contract_max_duration_days' in ml_dataset.columns and 'contract_min_duration_days' in ml_dataset.columns:
    ml_dataset['contract_duration_variance'] = (
        ml_dataset['contract_max_duration_days'] - ml_dataset['contract_min_duration_days']
    )
    print("   âœ“ contract_duration_variance: Variety in contract lengths")

# Service intensity per contract
ml_dataset['services_per_contract'] = ml_dataset['work_count'] / (ml_dataset['contract_count'] + 1)
print("   âœ“ services_per_contract: Average service visits per contract")


# ============================================================================
# 8. INTERACTION-SPECIFIC FEATURES
# ============================================================================
print("\n8ï¸âƒ£ Interaction-Specific Features")

# Complaint intensity per service
ml_dataset['complaints_per_service'] = ml_dataset['interaction_voc_count'] / (ml_dataset['work_count'] + 1)
print("   âœ“ complaints_per_service: VOC rate relative to service visits")

# Content engagement (has detailed content)
if 'interaction_has_ì ‘ìˆ˜ë‚´ìš©' in ml_dataset.columns:
    ml_dataset['has_detailed_interactions'] = ml_dataset['interaction_has_ì ‘ìˆ˜ë‚´ìš©']
    print("   âœ“ has_detailed_interactions: Indicator for detailed interaction records")

# Average interaction content length
if 'interaction_avg_length' in ml_dataset.columns:
    ml_dataset['interaction_detail_level'] = ml_dataset['interaction_avg_length']
    print("   âœ“ interaction_detail_level: Average content length of interactions")


# ============================================================================
# SUMMARY
# ============================================================================
new_feature_count = len(ml_dataset.columns)
added_features = new_feature_count - original_feature_count

print("\n" + "="*80)
print(f"âœ… Feature Engineering Complete!")
print("="*80)
print(f"   Original features: {original_feature_count - 3}")  # Exclude ID and target
print(f"   New derived features: {added_features}")
print(f"   Total features now: {new_feature_count - 3}")  # Exclude ID and target
print(f"   Dataset shape: {ml_dataset.shape}")

# Check for any NaN values in new features
new_cols = ml_dataset.columns[original_feature_count:]
nan_counts = ml_dataset[new_cols].isnull().sum()
features_with_nans = nan_counts[nan_counts > 0]

if len(features_with_nans) > 0:
    print(f"\nâš ï¸ New features with missing values ({len(features_with_nans)}):")
    for feat, count in features_with_nans.items():
        print(f"   - {feat}: {count} ({count/len(ml_dataset)*100:.1f}%)")
else:
    print(f"\nâœ“ No missing values in new features")

print("="*80)
````

````python
# Add time-windowed features and remove combined_risk_score

print("="*80)
print("ğŸ”§ Adding Time-Windowed Features & Removing Combined Risk Score")
print("="*80)

# First, let's extract more detailed temporal information from JSON columns
# We need the actual dates from interaction_history and work_history

def extract_interaction_dates(interaction_json):
    """Extract all interaction dates for time window analysis"""
    from datetime import datetime
    
    if pd.isna(interaction_json) or interaction_json == '' or interaction_json == '[]':
        return []
    
    try:
        if isinstance(interaction_json, str):
            interactions = json.loads(interaction_json)
        else:
            interactions = interaction_json
            
        if not isinstance(interactions, list):
            return []
        
        dates = []
        for item in interactions:
            date_str = item.get('ì ‘ìˆ˜ì¼ì‹œ', '')
            if date_str:
                try:
                    if 'T' in date_str:
                        date_obj = datetime.fromisoformat(date_str)
                    else:
                        date_obj = datetime.strptime(date_str, '%Y-%m-%d')
                    dates.append(date_obj)
                except:
                    pass
        return dates
    except:
        return []

def extract_work_dates(work_json):
    """Extract all work dates for time window analysis"""
    from datetime import datetime
    
    if pd.isna(work_json) or work_json == '' or work_json == '[]':
        return []
    
    try:
        if isinstance(work_json, str):
            works = json.loads(work_json)
        else:
            works = work_json
            
        if not isinstance(works, list):
            return []
        
        dates = []
        for item in works:
            # Try multiple date field names
            date_str = item.get('ì‘ì—…ì¼ì', item.get('ì‘ì—…ì¼ì‹œ', item.get('date', '')))
            if date_str:
                try:
                    if 'T' in str(date_str):
                        date_obj = datetime.fromisoformat(str(date_str))
                    else:
                        date_obj = datetime.strptime(str(date_str), '%Y-%m-%d')
                    dates.append(date_obj)
                except:
                    pass
        return dates
    except:
        return []

print("\n1ï¸âƒ£ Extracting interaction dates...")
interaction_dates = final_df_with_label['interaction_history'].apply(extract_interaction_dates)

print("2ï¸âƒ£ Extracting work dates...")
work_dates = final_df_with_label['ì‘ì—…ì´ë ¥'].apply(extract_work_dates)

# Now create time-windowed features
from datetime import datetime, timedelta

reference_date = datetime.now()

print("\n3ï¸âƒ£ Creating time-windowed features...")

# Interaction features for different time windows
time_windows = [30, 60, 90, 180, 365]

for days in time_windows:
    cutoff_date = reference_date - timedelta(days=days)
    
    # Count interactions in this window
    ml_dataset[f'interaction_last_{days}_days'] = interaction_dates.apply(
        lambda dates: sum(1 for d in dates if d >= cutoff_date)
    )
    
    # Count works in this window
    ml_dataset[f'work_last_{days}_days'] = work_dates.apply(
        lambda dates: sum(1 for d in dates if d >= cutoff_date)
    )
    
    print(f"   âœ“ Created features for last {days} days")

# Create ratio features (recent vs total activity)
print("\n4ï¸âƒ£ Creating activity ratio features...")

ml_dataset['interaction_ratio_30_to_total'] = ml_dataset['interaction_last_30_days'] / (ml_dataset['interaction_count'] + 1)
ml_dataset['interaction_ratio_90_to_total'] = ml_dataset['interaction_last_90_days'] / (ml_dataset['interaction_count'] + 1)
ml_dataset['work_ratio_30_to_total'] = ml_dataset['work_last_30_days'] / (ml_dataset['work_count'] + 1)
ml_dataset['work_ratio_90_to_total'] = ml_dataset['work_last_90_days'] / (ml_dataset['work_count'] + 1)

print("   âœ“ interaction_ratio_30_to_total")
print("   âœ“ interaction_ratio_90_to_total")
print("   âœ“ work_ratio_30_to_total")
print("   âœ“ work_ratio_90_to_total")

# Trend indicators (comparing recent windows)
print("\n5ï¸âƒ£ Creating trend indicators...")

ml_dataset['interaction_trend_30_vs_60'] = ml_dataset['interaction_last_30_days'] / (ml_dataset['interaction_last_60_days'] - ml_dataset['interaction_last_30_days'] + 1)
ml_dataset['interaction_trend_60_vs_90'] = ml_dataset['interaction_last_60_days'] / (ml_dataset['interaction_last_90_days'] - ml_dataset['interaction_last_60_days'] + 1)
ml_dataset['work_trend_30_vs_60'] = ml_dataset['work_last_30_days'] / (ml_dataset['work_last_60_days'] - ml_dataset['work_last_30_days'] + 1)
ml_dataset['work_trend_60_vs_90'] = ml_dataset['work_last_60_days'] / (ml_dataset['work_last_90_days'] - ml_dataset['work_last_60_days'] + 1)

print("   âœ“ interaction_trend_30_vs_60 (recent activity acceleration)")
print("   âœ“ interaction_trend_60_vs_90")
print("   âœ“ work_trend_30_vs_60")
print("   âœ“ work_trend_60_vs_90")

# Activity decay indicators
print("\n6ï¸âƒ£ Creating activity decay indicators...")

ml_dataset['interaction_decay_score'] = (
    ml_dataset['interaction_last_30_days'] * 1.0 +
    ml_dataset['interaction_last_60_days'] * 0.5 +
    ml_dataset['interaction_last_90_days'] * 0.25
) / (ml_dataset['interaction_count'] + 1)

ml_dataset['work_decay_score'] = (
    ml_dataset['work_last_30_days'] * 1.0 +
    ml_dataset['work_last_60_days'] * 0.5 +
    ml_dataset['work_last_90_days'] * 0.25
) / (ml_dataset['work_count'] + 1)

print("   âœ“ interaction_decay_score (weighted recent activity)")
print("   âœ“ work_decay_score")

# Remove combined_risk_score
print("\n7ï¸âƒ£ Removing combined_risk_score...")
if 'combined_risk_score' in ml_dataset.columns:
    ml_dataset = ml_dataset.drop(columns=['combined_risk_score'])
    print("   âœ“ Removed combined_risk_score")
else:
    print("   âš ï¸ combined_risk_score not found in dataset")

# Summary
print("\n" + "="*80)
print("âœ… Feature Engineering Update Complete!")
print("="*80)
print(f"Dataset shape: {ml_dataset.shape}")
print(f"\nAdded features:")
print(f"   - Time-windowed counts (10): interaction/work for 30/60/90/180/365 days")
print(f"   - Activity ratios (4): Recent activity vs total")
print(f"   - Trend indicators (4): Comparing adjacent time windows")
print(f"   - Decay scores (2): Weighted recent activity")
print(f"   Total new features: 20")
print(f"\nRemoved features:")
print(f"   - combined_risk_score")
print(f"\nNet change: +19 features")
print("="*80)
````

````python
# Feature extraction function for purchase logs (to be used after data loads)

def extract_purchase_features(purchase_json):
    """Extract features from purchase_logs JSON"""
    from datetime import datetime
    import numpy as np
    
    features = {}
    
    if pd.isna(purchase_json) or purchase_json == '' or purchase_json == '[]':
        # No purchase data
        features['purchase_count'] = 0
        features['purchase_total_quantity'] = 0
        features['purchase_unique_items'] = 0
        features['purchase_unique_orders'] = 0
        features['purchase_days_since_last'] = None
        features['purchase_days_since_first'] = None
        features['purchase_frequency_days'] = None
        features['purchase_avg_quantity_per_order'] = 0
        features['purchase_buy_kind_diversity'] = 0
        features['purchase_channel_diversity'] = 0
        features['purchase_most_common_channel'] = None
        features['purchase_last_30_days'] = 0
        features['purchase_last_60_days'] = 0
        features['purchase_last_90_days'] = 0
        features['purchase_last_180_days'] = 0
        return features
    
    try:
        # Parse JSON
        if isinstance(purchase_json, str):
            purchases = json.loads(purchase_json)
        else:
            purchases = purchase_json
            
        if not isinstance(purchases, list) or len(purchases) == 0:
            # Empty purchases
            features['purchase_count'] = 0
            features['purchase_total_quantity'] = 0
            features['purchase_unique_items'] = 0
            features['purchase_unique_orders'] = 0
            features['purchase_days_since_last'] = None
            features['purchase_days_since_first'] = None
            features['purchase_frequency_days'] = None
            features['purchase_avg_quantity_per_order'] = 0
            features['purchase_buy_kind_diversity'] = 0
            features['purchase_channel_diversity'] = 0
            features['purchase_most_common_channel'] = None
            features['purchase_last_30_days'] = 0
            features['purchase_last_60_days'] = 0
            features['purchase_last_90_days'] = 0
            features['purchase_last_180_days'] = 0
            return features
        
        # Basic counts
        features['purchase_count'] = len(purchases)
        
        # Quantity metrics
        quantities = [int(p.get('quantity', 0)) for p in purchases if p.get('quantity')]
        features['purchase_total_quantity'] = sum(quantities)
        
        # Item diversity
        unique_items = set(p.get('item_code') for p in purchases if p.get('item_code'))
        features['purchase_unique_items'] = len(unique_items)
        
        # Order diversity
        unique_orders = set(p.get('order_code') for p in purchases if p.get('order_code'))
        features['purchase_unique_orders'] = len(unique_orders)
        
        # Average quantity per order
        if len(unique_orders) > 0:
            features['purchase_avg_quantity_per_order'] = features['purchase_total_quantity'] / len(unique_orders)
        else:
            features['purchase_avg_quantity_per_order'] = 0
        
        # Buy kind diversity
        buy_kinds = set(p.get('buy_kind') for p in purchases if p.get('buy_kind'))
        features['purchase_buy_kind_diversity'] = len(buy_kinds)
        
        # Channel diversity
        channels = [p.get('sale_channel_code') for p in purchases if p.get('sale_channel_code')]
        unique_channels = set(channels)
        features['purchase_channel_diversity'] = len(unique_channels)
        
        # Most common channel
        if channels:
            from collections import Counter
            most_common = Counter(channels).most_common(1)[0][0]
            features['purchase_most_common_channel'] = most_common
        else:
            features['purchase_most_common_channel'] = None
        
        # Date-based features
        dates = []
        for p in purchases:
            date_str = p.get('purchase_date', '')
            if date_str:
                try:
                    # Parse date (YYYY-MM-DD format)
                    date_obj = datetime.strptime(str(date_str), '%Y-%m-%d')
                    dates.append(date_obj)
                except:
                    pass
        
        if dates:
            dates_sorted = sorted(dates)
            most_recent = dates_sorted[-1]
            oldest = dates_sorted[0]
            
            # Days since last and first purchase
            now = datetime.now()
            features['purchase_days_since_last'] = (now - most_recent).days
            features['purchase_days_since_first'] = (now - oldest).days
            
            # Purchase frequency
            if len(dates) > 1:
                date_span = (most_recent - oldest).days
                features['purchase_frequency_days'] = date_span / (len(dates) - 1)
            else:
                features['purchase_frequency_days'] = 0
            
            # Time-windowed counts
            from datetime import timedelta
            cutoff_30 = now - timedelta(days=30)
            cutoff_60 = now - timedelta(days=60)
            cutoff_90 = now - timedelta(days=90)
            cutoff_180 = now - timedelta(days=180)
            
            features['purchase_last_30_days'] = sum(1 for d in dates if d >= cutoff_30)
            features['purchase_last_60_days'] = sum(1 for d in dates if d >= cutoff_60)
            features['purchase_last_90_days'] = sum(1 for d in dates if d >= cutoff_90)
            features['purchase_last_180_days'] = sum(1 for d in dates if d >= cutoff_180)
        else:
            features['purchase_days_since_last'] = None
            features['purchase_days_since_first'] = None
            features['purchase_frequency_days'] = None
            features['purchase_last_30_days'] = 0
            features['purchase_last_60_days'] = 0
            features['purchase_last_90_days'] = 0
            features['purchase_last_180_days'] = 0
            
    except Exception as e:
        # Return default values on error
        features['purchase_count'] = 0
        features['purchase_total_quantity'] = 0
        features['purchase_unique_items'] = 0
        features['purchase_unique_orders'] = 0
        features['purchase_days_since_last'] = None
        features['purchase_days_since_first'] = None
        features['purchase_frequency_days'] = None
        features['purchase_avg_quantity_per_order'] = 0
        features['purchase_buy_kind_diversity'] = 0
        features['purchase_channel_diversity'] = 0
        features['purchase_most_common_channel'] = None
        features['purchase_last_30_days'] = 0
        features['purchase_last_60_days'] = 0
        features['purchase_last_90_days'] = 0
        features['purchase_last_180_days'] = 0
    
    return features

print("âœ… Purchase feature extraction function defined")
print("   Features to be extracted:")
print("   - purchase_count: Total number of purchases")
print("   - purchase_total_quantity: Sum of all quantities")
print("   - purchase_unique_items: Number of different items purchased")
print("   - purchase_unique_orders: Number of unique orders")
print("   - purchase_days_since_last: Days since most recent purchase")
print("   - purchase_days_since_first: Days since first purchase")
print("   - purchase_frequency_days: Average days between purchases")
print("   - purchase_avg_quantity_per_order: Average quantity per order")
print("   - purchase_buy_kind_diversity: Number of different buy kinds")
print("   - purchase_channel_diversity: Number of different sales channels")
print("   - purchase_most_common_channel: Most frequently used channel")
print("   - purchase_last_30/60/90/180_days: Time-windowed purchase counts")
````

````python
# Re-define purchase feature extraction function with proper imports

import pandas as pd
import json
from datetime import datetime, timedelta
import numpy as np

def extract_purchase_features(purchase_json):
    """Extract features from purchase_logs JSON"""
    
    features = {}
    
    if pd.isna(purchase_json) or purchase_json == '' or purchase_json == '[]':
        # No purchase data
        features['purchase_count'] = 0
        features['purchase_total_quantity'] = 0
        features['purchase_unique_items'] = 0
        features['purchase_unique_orders'] = 0
        features['purchase_days_since_last'] = None
        features['purchase_days_since_first'] = None
        features['purchase_frequency_days'] = None
        features['purchase_avg_quantity_per_order'] = 0
        features['purchase_buy_kind_diversity'] = 0
        features['purchase_channel_diversity'] = 0
        features['purchase_most_common_channel'] = None
        features['purchase_last_30_days'] = 0
        features['purchase_last_60_days'] = 0
        features['purchase_last_90_days'] = 0
        features['purchase_last_180_days'] = 0
        return features
    
    try:
        # Parse JSON
        if isinstance(purchase_json, str):
            purchases = json.loads(purchase_json)
        else:
            purchases = purchase_json
            
        if not isinstance(purchases, list) or len(purchases) == 0:
            # Empty purchases
            features['purchase_count'] = 0
            features['purchase_total_quantity'] = 0
            features['purchase_unique_items'] = 0
            features['purchase_unique_orders'] = 0
            features['purchase_days_since_last'] = None
            features['purchase_days_since_first'] = None
            features['purchase_frequency_days'] = None
            features['purchase_avg_quantity_per_order'] = 0
            features['purchase_buy_kind_diversity'] = 0
            features['purchase_channel_diversity'] = 0
            features['purchase_most_common_channel'] = None
            features['purchase_last_30_days'] = 0
            features['purchase_last_60_days'] = 0
            features['purchase_last_90_days'] = 0
            features['purchase_last_180_days'] = 0
            return features
        
        # Basic counts
        features['purchase_count'] = len(purchases)
        
        # Quantity metrics
        quantities = [int(p.get('quantity', 0)) for p in purchases if p.get('quantity')]
        features['purchase_total_quantity'] = sum(quantities)
        
        # Item diversity
        unique_items = set(p.get('item_code') for p in purchases if p.get('item_code'))
        features['purchase_unique_items'] = len(unique_items)
        
        # Order diversity
        unique_orders = set(p.get('order_code') for p in purchases if p.get('order_code'))
        features['purchase_unique_orders'] = len(unique_orders)
        
        # Average quantity per order
        if len(unique_orders) > 0:
            features['purchase_avg_quantity_per_order'] = features['purchase_total_quantity'] / len(unique_orders)
        else:
            features['purchase_avg_quantity_per_order'] = 0
        
        # Buy kind diversity
        buy_kinds = set(p.get('buy_kind') for p in purchases if p.get('buy_kind'))
        features['purchase_buy_kind_diversity'] = len(buy_kinds)
        
        # Channel diversity
        channels = [p.get('sale_channel_code') for p in purchases if p.get('sale_channel_code')]
        unique_channels = set(channels)
        features['purchase_channel_diversity'] = len(unique_channels)
        
        # Most common channel
        if channels:
            from collections import Counter
            most_common = Counter(channels).most_common(1)[0][0]
            features['purchase_most_common_channel'] = most_common
        else:
            features['purchase_most_common_channel'] = None
        
        # Date-based features
        dates = []
        for p in purchases:
            date_str = p.get('purchase_date', '')
            if date_str:
                try:
                    # Parse date (YYYY-MM-DD format)
                    date_obj = datetime.strptime(str(date_str), '%Y-%m-%d')
                    dates.append(date_obj)
                except:
                    pass
        
        if dates:
            dates_sorted = sorted(dates)
            most_recent = dates_sorted[-1]
            oldest = dates_sorted[0]
            
            # Days since last and first purchase
            now = datetime.now()
            features['purchase_days_since_last'] = (now - most_recent).days
            features['purchase_days_since_first'] = (now - oldest).days
            
            # Purchase frequency
            if len(dates) > 1:
                date_span = (most_recent - oldest).days
                features['purchase_frequency_days'] = date_span / (len(dates) - 1)
            else:
                features['purchase_frequency_days'] = 0
            
            # Time-windowed counts
            cutoff_30 = now - timedelta(days=30)
            cutoff_60 = now - timedelta(days=60)
            cutoff_90 = now - timedelta(days=90)
            cutoff_180 = now - timedelta(days=180)
            
            features['purchase_last_30_days'] = sum(1 for d in dates if d >= cutoff_30)
            features['purchase_last_60_days'] = sum(1 for d in dates if d >= cutoff_60)
            features['purchase_last_90_days'] = sum(1 for d in dates if d >= cutoff_90)
            features['purchase_last_180_days'] = sum(1 for d in dates if d >= cutoff_180)
        else:
            features['purchase_days_since_last'] = None
            features['purchase_days_since_first'] = None
            features['purchase_frequency_days'] = None
            features['purchase_last_30_days'] = 0
            features['purchase_last_60_days'] = 0
            features['purchase_last_90_days'] = 0
            features['purchase_last_180_days'] = 0
            
    except Exception as e:
        # Return default values on error
        features['purchase_count'] = 0
        features['purchase_total_quantity'] = 0
        features['purchase_unique_items'] = 0
        features['purchase_unique_orders'] = 0
        features['purchase_days_since_last'] = None
        features['purchase_days_since_first'] = None
        features['purchase_frequency_days'] = None
        features['purchase_avg_quantity_per_order'] = 0
        features['purchase_buy_kind_diversity'] = 0
        features['purchase_channel_diversity'] = 0
        features['purchase_most_common_channel'] = None
        features['purchase_last_30_days'] = 0
        features['purchase_last_60_days'] = 0
        features['purchase_last_90_days'] = 0
        features['purchase_last_180_days'] = 0
    
    return features

print("âœ… Purchase feature extraction function re-defined with proper imports")
````

## Step 1: Data Preprocessing

Prepare the data for modeling:
- Separate features from target
- Handle missing values strategically
- Split into train/validation/test sets with stratification
- Scale features

````python
# Separate features, target, and metadata
metadata_cols = ['ê³ ê°ì½”ë“œ', 'ê³ ê°ëª…', 'ìµœê·¼_í•´ì•½ì¼ì_SM', 'ìµœì´ˆ_í•´ì•½ì¼ì_SM']
target_col = 'í•´ì•½ì—¬ë¶€'

feature_cols = [col for col in ml_dataset.columns 
                if col not in metadata_cols + [target_col]]

X = ml_dataset[feature_cols].copy()
y = ml_dataset[target_col].copy()
customer_ids = ml_dataset['ê³ ê°ì½”ë“œ'].copy()

print(f"âœ… Data Separated:")
print(f"   Features (X): {X.shape}")
print(f"   Target (y): {y.shape}")
print(f"   Feature count: {len(feature_cols)}")
print(f"\nğŸ“Š Target Distribution:")
print(y.value_counts())
print(f"   Churn Rate: {y.mean()*100:.2f}%")
````

````python
# Analyze missing values
print("ğŸ” Missing Value Analysis:")
print("=" * 80)

missing_stats = pd.DataFrame({
    'Missing_Count': X.isnull().sum(),
    'Missing_Pct': (X.isnull().sum() / len(X) * 100).round(2)
}).sort_values('Missing_Pct', ascending=False)

# Show features with missing values
missing_features = missing_stats[missing_stats['Missing_Count'] > 0]
print(f"\nFeatures with missing values: {len(missing_features)} / {len(feature_cols)}")
print(f"\nTop 20 features with most missing data:")
print(missing_features.head(20))

# Categorize by missing percentage
high_missing = missing_features[missing_features['Missing_Pct'] > 50]
medium_missing = missing_features[(missing_features['Missing_Pct'] > 5) & (missing_features['Missing_Pct'] <= 50)]
low_missing = missing_features[missing_features['Missing_Pct'] <= 5]

print(f"\nğŸ“Š Missing Value Categories:")
print(f"   High (>50%): {len(high_missing)} features")
print(f"   Medium (5-50%): {len(medium_missing)} features")
print(f"   Low (<5%): {len(low_missing)} features")
````

````python
# Handle missing values strategically
from sklearn.impute import SimpleImputer

print("ğŸ”§ Handling Missing Values...")
print("=" * 80)

X_imputed = X.copy()

# Strategy 1: For recency features (days_since_last), use a high value (no recent activity)
recency_cols = [col for col in X.columns if 'days_since_last' in col]
print(f"\n1ï¸âƒ£ Recency features (impute with 9999): {len(recency_cols)}")
for col in recency_cols:
    missing_count = X_imputed[col].isnull().sum()
    if missing_count > 0:
        X_imputed[col].fillna(9999, inplace=True)
        print(f"   âœ“ {col}: {missing_count} nulls â†’ 9999")

# Strategy 2: For count/frequency features, use 0 (no activity)
count_cols = [col for col in X.columns if any(word in col for word in ['count', 'frequency', 'ratio', 'last_30', 'last_60', 'last_90', 'last_180', 'last_365'])]
count_cols = [col for col in count_cols if col not in recency_cols]
print(f"\n2ï¸âƒ£ Count/frequency features (impute with 0): {len(count_cols)}")
imputed_counts = 0
for col in count_cols:
    missing_count = X_imputed[col].isnull().sum()
    if missing_count > 0:
        X_imputed[col].fillna(0, inplace=True)
        imputed_counts += 1
print(f"   âœ“ Imputed {imputed_counts} features with 0")

# Strategy 3: For score/quality features, use median
score_cols = [col for col in X.columns if any(word in col for word in ['score', 'csi', 'quality'])]
score_cols = [col for col in score_cols if col not in recency_cols + count_cols]
print(f"\n3ï¸âƒ£ Score/quality features (impute with median): {len(score_cols)}")
for col in score_cols:
    missing_count = X_imputed[col].isnull().sum()
    if missing_count > 0:
        median_val = X_imputed[col].median()
        X_imputed[col].fillna(median_val, inplace=True)
        print(f"   âœ“ {col}: {missing_count} nulls â†’ {median_val:.2f}")

# Strategy 4: For remaining numeric features, use median
remaining_numeric = X_imputed.select_dtypes(include=[np.number]).columns
remaining_with_nulls = [col for col in remaining_numeric if X_imputed[col].isnull().sum() > 0]
print(f"\n4ï¸âƒ£ Remaining numeric features (impute with median): {len(remaining_with_nulls)}")
if len(remaining_with_nulls) > 0:
    imputer = SimpleImputer(strategy='median')
    X_imputed[remaining_with_nulls] = imputer.fit_transform(X_imputed[remaining_with_nulls])
    print(f"   âœ“ Imputed {len(remaining_with_nulls)} features")

# Verify no missing values remain
remaining_nulls = X_imputed.isnull().sum().sum()
print(f"\nâœ… Missing values after imputation: {remaining_nulls}")

if remaining_nulls > 0:
    print(f"âš ï¸ Still have {remaining_nulls} nulls in:")
    print(X_imputed.isnull().sum()[X_imputed.isnull().sum() > 0])
````

````python
# Train/Validation/Test Split with Stratification
from sklearn.model_selection import train_test_split

print("ğŸ“Š Splitting Data (Stratified)...")
print("=" * 80)

# First split: 70% train+val, 30% test
X_temp, X_test, y_temp, y_test, ids_temp, ids_test = train_test_split(
    X_imputed, y, customer_ids, 
    test_size=0.30, 
    random_state=42, 
    stratify=y
)

# Second split: 70% train (49% of total), 30% validation (21% of total)
X_train, X_val, y_train, y_val, ids_train, ids_val = train_test_split(
    X_temp, y_temp, ids_temp,
    test_size=0.30,
    random_state=42,
    stratify=y_temp
)

print(f"\nâœ… Split Complete:")
print(f"   Train: {X_train.shape[0]:,} samples ({X_train.shape[0]/len(X)*100:.1f}%)")
print(f"   Val:   {X_val.shape[0]:,} samples ({X_val.shape[0]/len(X)*100:.1f}%)")
print(f"   Test:  {X_test.shape[0]:,} samples ({X_test.shape[0]/len(X)*100:.1f}%)")

print(f"\nğŸ“ˆ Churn Distribution (Stratification Check):")
print(f"   Train: {y_train.mean()*100:.2f}% churn ({y_train.sum()} / {len(y_train)})")
print(f"   Val:   {y_val.mean()*100:.2f}% churn ({y_val.sum()} / {len(y_val)})")
print(f"   Test:  {y_test.mean()*100:.2f}% churn ({y_test.sum()} / {len(y_test)})")
print(f"   Original: {y.mean()*100:.2f}% churn")

print(f"\nâœ… Class distribution maintained across all splits!")
````

## Step 2: Baseline Model - XGBoost

Train an XGBoost classifier with class weights to handle imbalance:
- No feature scaling needed (tree-based model)
- Use `scale_pos_weight` for class imbalance
- Evaluate on validation set

````python
# Install XGBoost if needed
try:
    import xgboost as xgb
    print("âœ… XGBoost already installed")
except ImportError:
    print("ğŸ“¦ Installing XGBoost...")
    import subprocess
    subprocess.check_call(['pip', 'install', 'xgboost'])
    import xgboost as xgb
    print("âœ… XGBoost installed successfully")
````

````python
# Train XGBoost with class weight handling
from sklearn.metrics import (classification_report, confusion_matrix, 
                             roc_auc_score, roc_curve, precision_recall_curve,
                             average_precision_score, f1_score)
import time

print("ğŸš€ Training XGBoost Classifier...")
print("=" * 80)

# Calculate scale_pos_weight for class imbalance
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()
print(f"\nâš–ï¸ Class Balance:")
print(f"   Negative samples: {(y_train == 0).sum()}")
print(f"   Positive samples: {(y_train == 1).sum()}")
print(f"   Scale pos weight: {scale_pos_weight:.2f}")

# XGBoost parameters
params = {
    'objective': 'binary:logistic',
    'eval_metric': ['logloss', 'auc'],
    'scale_pos_weight': scale_pos_weight,
    'max_depth': 6,
    'learning_rate': 0.1,
    'n_estimators': 100,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'random_state': 42,
    'tree_method': 'hist',
    'early_stopping_rounds': 10
}

print(f"\nğŸ“ Model Parameters:")
for key, value in params.items():
    print(f"   {key}: {value}")

# Train model
start_time = time.time()
model = xgb.XGBClassifier(**params)

model.fit(
    X_train, y_train,
    eval_set=[(X_train, y_train), (X_val, y_val)],
    verbose=True
)

training_time = time.time() - start_time
print(f"\nâœ… Training Complete!")
print(f"   Time: {training_time:.2f} seconds")
print(f"   Best iteration: {model.best_iteration}")
print(f"   Best score: {model.best_score:.4f}")
````

````python
# Make predictions on validation set
print("ğŸ¯ Making Predictions on Validation Set...")
print("=" * 80)

y_val_pred_proba = model.predict_proba(X_val)[:, 1]
y_val_pred = model.predict(X_val)

# Calculate metrics
val_auc_roc = roc_auc_score(y_val, y_val_pred_proba)
val_avg_precision = average_precision_score(y_val, y_val_pred_proba)
val_f1 = f1_score(y_val, y_val_pred)

print(f"\nğŸ“Š Validation Set Performance:")
print(f"   AUC-ROC:          {val_auc_roc:.4f}")
print(f"   Average Precision: {val_avg_precision:.4f}")
print(f"   F1-Score:         {val_f1:.4f}")

print(f"\nğŸ“‹ Classification Report:")
print(classification_report(y_val, y_val_pred, target_names=['Retained', 'Churned']))

print(f"\nğŸ” Confusion Matrix:")
cm = confusion_matrix(y_val, y_val_pred)
print(f"                Predicted")
print(f"              Retained  Churned")
print(f"Actual Retained    {cm[0,0]:3d}      {cm[0,1]:3d}")
print(f"       Churned     {cm[1,0]:3d}      {cm[1,1]:3d}")

tn, fp, fn, tp = cm.ravel()
print(f"\nğŸ“ˆ Detailed Metrics:")
print(f"   True Negatives:  {tn} (correctly predicted retained)")
print(f"   False Positives: {fp} (predicted churn but retained)")
print(f"   False Negatives: {fn} (predicted retained but churned)")
print(f"   True Positives:  {tp} (correctly predicted churn)")
print(f"\n   Specificity: {tn/(tn+fp):.4f} (correctly identify retained)")
print(f"   Sensitivity: {tp/(tp+fn):.4f} (correctly identify churned)")
````

````python
# Visualize ROC and Precision-Recall Curves
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# ROC Curve
fpr, tpr, thresholds_roc = roc_curve(y_val, y_val_pred_proba)
ax1.plot(fpr, tpr, linewidth=2, label=f'XGBoost (AUC = {val_auc_roc:.4f})')
ax1.plot([0, 1], [0, 1], 'k--', linewidth=1, label='Random Classifier')
ax1.set_xlabel('False Positive Rate', fontsize=12)
ax1.set_ylabel('True Positive Rate', fontsize=12)
ax1.set_title('ROC Curve - Validation Set', fontsize=14, fontweight='bold')
ax1.legend(fontsize=10)
ax1.grid(alpha=0.3)

# Precision-Recall Curve
precision, recall, thresholds_pr = precision_recall_curve(y_val, y_val_pred_proba)
ax2.plot(recall, precision, linewidth=2, label=f'XGBoost (AP = {val_avg_precision:.4f})')
ax2.axhline(y=y_val.mean(), color='k', linestyle='--', linewidth=1, 
            label=f'Baseline ({y_val.mean():.4f})')
ax2.set_xlabel('Recall', fontsize=12)
ax2.set_ylabel('Precision', fontsize=12)
ax2.set_title('Precision-Recall Curve - Validation Set', fontsize=14, fontweight='bold')
ax2.legend(fontsize=10)
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("âœ… ROC-AUC shows how well the model distinguishes between classes")
print("âœ… PR-AUC is better for imbalanced datasets (focuses on positive class)")
````

## Step 3: Final Evaluation on Test Set

Evaluate the model on the held-out test set to get unbiased performance estimates.
