## 해약 여부 라벨 생성 (Based on 해약여부관련_정리_v0_2.ipynb)

````python
# 해약 여부 라벨을 생성하기 위한 데이터 확인
# final_df에서 필요한 컬럼 확인
print("Available columns in final_df:")
print(final_df.columns.tolist())
print(f"\nShape: {final_df.shape}")
print(f"\nSample data:")
final_df.head()
````

````python
# Step 1: 계약 데이터 조회 (신계약 마스터 + 상세 + 해약 정보)
# 고객코드 리스트 생성
custcode_list = ','.join([f"'{code}'" for code in final_df['고객코드'].unique()])

# 계약 정보 조회 쿼리 (해약여부관련_정리_v0_2.ipynb의 로직 참조)
contract_query = f"""
SELECT 
    A.고객코드,
    A.계약일련번호,
    A.계약일자,
    B.계약대상,
    B.시작일자,
    B.종료일자,
    B.해약일련번호,
    Cust.고객명 as 고객명_CU,
    Cust.고객구분 as 고객구분_CU,
    D.해약일자 as 해약일자_SM,
    m.Gerldesp,
    m.CONTDIVS as 계약대상_중분류,
    m.M_nm as 계약대상_중분류명,
    m.CONTLATY as 계약대상_대분류,
    m.L_nm as 계약대상_대분류명
FROM CESCOEIS.dbo.TB_신계약_마스타 A
LEFT JOIN CESCOEIS.dbo.TB_신계약_상세 B
ON A.계약일련번호 = B.계약일련번호
LEFT JOIN CESCOEIS.dbo.TB_신계약_마스타_해약 C
ON A.계약일련번호 = C.계약일련번호
LEFT JOIN CESCOEIS.dbo.TB_신계약_부분해약 D
ON B.해약일련번호 = D.해약일련번호
LEFT JOIN (
    SELECT 고객코드, 고객명, 고객구분, 본사코드, 등록일자 FROM CESCOEIS.dbo.TB_고객
    UNION ALL
    SELECT 고객코드, 고객명, 고객구분, 본사코드, 등록일자 FROM CESCOEIS.dbo.TB_단독고객
) Cust
ON A.고객코드 = Cust.고객코드
LEFT JOIN (
    SELECT MK.Gerlcode, MK.Gerlgubn, MK.Gerldesp
            , CD.CONTDIVS, M.NAME as M_nm
            , CD.CONTLATY, L.NAME as L_nm
    FROM CESCOEIS.dbo.MKTCGERL MK
    LEFT OUTER JOIN CESCOMIS.dbo.TB_CP_CONTRACTDETAIL_HIS CD
    ON MK.Gerlcode = CD.CONTTARG
    AND LEFT('20250501',6) BETWEEN CD.ST_YYYYMM AND CD.ED_YYYYMM
    LEFT OUTER JOIN CESCOMIS.dbo.TB_CODE L
    ON CD.CONTLATY = L.CODE
    LEFT OUTER JOIN CESCOMIS.dbo.TB_CODE M
    ON CD.CONTDIVS = M.CODE
    WHERE Gerlgubn IN ('계약대상','가정집계약대상')
) m
ON B.계약대상 = m.Gerlcode
WHERE B.계약일련번호 IS NOT NULL
AND A.고객코드 IN ({custcode_list})
AND A.고객코드 IS NOT NULL
AND A.고객코드 != ''
AND A.고객코드 != 'AT7728'
"""

print("Fetching contract data with joins...")
print(f"Number of customers: {final_df['고객코드'].nunique()}")
contract_df = rodb.execute_query(contract_query)

if contract_df is not None:
    print(f"✅ Contract data fetched: {len(contract_df)} records")
    display(contract_df.head())
else:
    print("❌ Failed to fetch contract data")
````

````python
# 테이블 컬럼 확인
column_query = """
SELECT TOP 5 * 
FROM CESCOEIS.dbo.TB_신계약_마스타
"""
print("Checking table columns...")
sample_data = rodb.execute_query(column_query)
if sample_data is not None:
    print(f"Available columns: {sample_data.columns.tolist()}")
    display(sample_data)
````

````python
# Step 2: 데이터 전처리 및 논리적 계약해지일 계산
import numpy as np

# 날짜 컬럼 전처리
contract_df['시작일자'] = pd.to_datetime(contract_df['시작일자'], errors='coerce')
contract_df['종료일자'] = pd.to_datetime(contract_df['종료일자'], errors='coerce')
contract_df['해약일자_SM'] = pd.to_datetime(contract_df['해약일자_SM'], errors='coerce')

# 논리적 계약해지일 계산 (해약일자가 있으면 해약일자, 없으면 종료일자)
contract_df['논리_계약해지일'] = contract_df['해약일자_SM'].fillna(contract_df['종료일자'])

# 유지일수 계산
contract_df['유지일수'] = (contract_df['논리_계약해지일'] - contract_df['시작일자']).dt.days

print(f"Processed {len(contract_df)} contract records")
print(f"\n해약일자_SM이 있는 건수: {contract_df['해약일자_SM'].notna().sum()}")
print(f"해약일자_SM이 없는 건수: {contract_df['해약일자_SM'].isna().sum()}")
contract_df[['고객코드', '계약일련번호', '계약대상', '시작일자', '종료일자', '해약일자_SM', '논리_계약해지일', '유지일수']].head(10)
````

````python
# Step 3: 계약대상별 집계 (고객코드 + 계약대상)
# 해약여부관련_정리_v0_2.ipynb의 로직 참조

contract_summary = contract_df.groupby(['고객코드','계약대상']).agg({
    '계약일련번호': ['count','nunique'],
    '시작일자': 'min',
    '종료일자': 'max',
    '해약일자_SM': ['count','min','max'],
    '논리_계약해지일': ['max','min'],
    '유지일수': 'sum'
}).reset_index()

# 컬럼명 단순화
contract_summary.columns = ['고객코드', '계약대상', '계약일련번호_cnt', '계약일련번호_unicnt', 
                            '최초계약_시작일자', '최근계약_종료일자',
                            '해약일자_SM_cnt', '최초_해약일자_SM', '최근_해약일자_SM', 
                            '최근_논리_계약해지일', '최초_논리_계약해지일', '유지일수_sum']

# 계약대상별 해약여부 계산: (총 계약 건수 - 해약 건수) > 0 이면 0(유지), 아니면 1(해약)
contract_summary['계약대상별_해약여부'] = np.where(
    (contract_summary['계약일련번호_cnt'] - contract_summary['해약일자_SM_cnt']) > 0, 
    0,  # 유지 중인 계약이 있음
    1   # 모든 계약이 해약됨
)

print(f"계약대상별 집계 완료: {len(contract_summary)} records")
print(f"\n계약대상별_해약여부 분포:")
print(contract_summary['계약대상별_해약여부'].value_counts())
contract_summary.head(10)
````

````python
# Step 4: 계약대상 분류 정보 merge
# 계약대상_대분류명 정보 추가

conttarg_info = contract_df[['계약대상', 'Gerldesp', '계약대상_중분류명', '계약대상_대분류명']].drop_duplicates()
print(f"계약대상 분류 정보: {len(conttarg_info)} unique contract targets")

contract_summary_with_class = pd.merge(
    contract_summary, 
    conttarg_info, 
    on='계약대상', 
    how='left'
)

# 계약대상_대분류명이 없는 건은 제외 (예전 계약건)
contract_summary_filtered = contract_summary_with_class[
    contract_summary_with_class['계약대상_대분류명'].notna()
].copy()

print(f"\n필터링 후: {len(contract_summary_filtered)} records")
print(f"제외된 건수: {len(contract_summary_with_class) - len(contract_summary_filtered)}")
print(f"\n계약대상_대분류명 분포:")
print(contract_summary_filtered['계약대상_대분류명'].value_counts())
contract_summary_filtered.head()
````

````python
# Step 5: 계약대상 대분류별 집계 (고객코드 + 계약대상_대분류명)
# 대분류별로 한 건이라도 유지 중인 계약이 있으면 0(유지), 모두 해약이면 1(해약)

contract_by_category = contract_summary_filtered.groupby(['고객코드','계약대상_대분류명']).agg({
    '계약대상': ['nunique', 'count'],
    '최초계약_시작일자': 'min',
    '최근계약_종료일자': 'max',
    '최근_해약일자_SM': 'max',
    '최초_해약일자_SM': 'min',
    '해약일자_SM_cnt': 'sum',
    '계약일련번호_unicnt': 'max',
    '최근_논리_계약해지일': 'max',
    '최초_논리_계약해지일': 'min',
    '계약대상별_해약여부': 'sum',
}).reset_index()

# 컬럼명 단순화
contract_by_category.columns = ['고객코드', '계약대상_대분류명', '계약대상_nunique', '계약대상_count',
                                '최초계약_시작일자', '최근계약_종료일자', '최근_해약일자_SM',
                                '최초_해약일자_SM', '해약일자_SM_cnt', '계약일련번호_unicnt',
                                '최근_논리_계약해지일', '최초_논리_계약해지일', '계약대상별_해약여부_sum']

# 계약대상_대분류_해약여부: (계약대상 개수 - 해약된 계약대상 개수) > 0 이면 0(유지), 아니면 1(해약)
contract_by_category['계약대상_대분류_해약여부'] = np.where(
    (contract_by_category['계약대상_count'] - contract_by_category['계약대상별_해약여부_sum']) > 0,
    0,  # 대분류 내 유지 중인 계약대상이 있음
    1   # 대분류 내 모든 계약대상이 해약됨
)

print(f"계약대상 대분류별 집계: {len(contract_by_category)} records")
print(f"\n계약대상_대분류_해약여부 분포:")
print(contract_by_category['계약대상_대분류_해약여부'].value_counts())
contract_by_category.head(10)
````

````python
# Step 6: 최종 해약 여부 라벨 생성 (고객코드별)
# 모든 대분류가 해약되면 0(해약), 한 건이라도 유지 중이면 1(유지)

final_churn_label = contract_by_category.groupby('고객코드').agg({
    '계약대상_대분류명': 'count',  # 대분류 개수
    '계약대상_대분류_해약여부': 'sum'  # 해약된 대분류 개수
}).reset_index()

final_churn_label.columns = ['고객코드', '계약대상_대분류명_count', '계약대상_대분류_해약여부_sum']

# 최종계약상태: (대분류 개수 - 해약된 대분류 개수) > 0 이면 1(유지), 아니면 0(해약)
final_churn_label['최종계약상태'] = np.where(
    (final_churn_label['계약대상_대분류명_count'] - final_churn_label['계약대상_대분류_해약여부_sum']) > 0,
    1,  # 유지 (at least one active contract)
    0   # 해약 (all contracts cancelled)
)

# 해약 여부 라벨 생성 (1: 해약, 0: 유지)
final_churn_label['해약여부'] = 1 - final_churn_label['최종계약상태']

print(f"최종 해약 여부 라벨 생성 완료: {len(final_churn_label)} customers")
print(f"\n최종계약상태 분포:")
print(f"  1 (유지): {(final_churn_label['최종계약상태'] == 1).sum()}")
print(f"  0 (해약): {(final_churn_label['최종계약상태'] == 0).sum()}")
print(f"\n해약여부 분포:")
print(f"  1 (해약): {(final_churn_label['해약여부'] == 1).sum()}")
print(f"  0 (유지): {(final_churn_label['해약여부'] == 0).sum()}")
print(f"\n해약율: {final_churn_label['해약여부'].mean():.2%}")

final_churn_label.head(20)
````

````python
# Step 7: final_df에 해약여부 라벨 추가
final_df_with_label = pd.merge(
    final_df,
    final_churn_label[['고객코드', '해약여부', '최종계약상태']],
    on='고객코드',
    how='left'
)

# 라벨이 없는 고객 확인 (계약 정보가 없는 고객)
missing_label_count = final_df_with_label['해약여부'].isna().sum()
print(f"라벨 매핑 완료:")
print(f"  전체 고객 수: {len(final_df_with_label)}")
print(f"  라벨이 있는 고객 수: {(~final_df_with_label['해약여부'].isna()).sum()}")
print(f"  라벨이 없는 고객 수: {missing_label_count}")

if missing_label_count > 0:
    print(f"\n⚠️ 라벨이 없는 고객은 계약 정보가 없는 고객입니다.")
    # 라벨이 없는 경우 기본값 0(유지)으로 설정
    final_df_with_label['해약여부'] = final_df_with_label['해약여부'].fillna(0).astype(int)
    final_df_with_label['최종계약상태'] = final_df_with_label['최종계약상태'].fillna(1).astype(int)

print(f"\n최종 데이터셋:")
print(f"  Shape: {final_df_with_label.shape}")
print(f"  해약 고객 수: {(final_df_with_label['해약여부'] == 1).sum()}")
print(f"  유지 고객 수: {(final_df_with_label['해약여부'] == 0).sum()}")
print(f"  해약율: {final_df_with_label['해약여부'].mean():.2%}")

final_df_with_label[['고객코드', '고객명', '해약여부', '최종계약상태']].head(20)
````

## 📊 해약 예측 모델링 데이터 준비 완료

**해약 여부 라벨 정의:**
- `해약여부 = 1`: 모든 계약대상 대분류가 해약된 고객 (완전 해약)
- `해약여부 = 0`: 최소 1개 이상의 계약대상 대분류가 유지 중인 고객 (유지)

**라벨 생성 로직:** (해약여부관련_정리_v0_2.ipynb 기반)
1. 계약대상별 해약여부 계산
2. 계약대상 대분류별 해약여부 집계
3. 고객별 최종 해약여부 결정

````python
# 해약 예측 모델링을 위한 최종 데이터셋 통계
import matplotlib.pyplot as plt
import seaborn as sns

print("="*60)
print("📊 해약 예측 모델링 데이터셋 요약")
print("="*60)
print(f"\n1. 데이터셋 기본 정보:")
print(f"   - 총 고객 수: {len(final_df_with_label):,}명")
print(f"   - 전체 특성 수: {final_df_with_label.shape[1]}개")
print(f"   - 타겟 변수: 해약여부")

print(f"\n2. 타겟 분포:")
print(f"   - 해약 고객 (1): {(final_df_with_label['해약여부'] == 1).sum():,}명 ({final_df_with_label['해약여부'].mean():.2%})")
print(f"   - 유지 고객 (0): {(final_df_with_label['해약여부'] == 0).sum():,}명 ({(1-final_df_with_label['해약여부'].mean()):.2%})")

# 클래스 불균형 비율
churn_ratio = final_df_with_label['해약여부'].mean()
imbalance_ratio = (1-churn_ratio) / churn_ratio
print(f"   - 클래스 불균형 비율: 1:{imbalance_ratio:.1f} (유지:해약)")

print(f"\n3. 계약대상 대분류별 해약율:")
# 대분류별 해약율 분석을 위해 contract_by_category 활용
category_churn = contract_by_category.groupby('계약대상_대분류명').agg({
    '고객코드': 'count',
    '계약대상_대분류_해약여부': 'sum'
}).reset_index()
category_churn['해약율'] = category_churn['계약대상_대분류_해약여부'] / category_churn['고객코드']
category_churn = category_churn.sort_values('해약율', ascending=False)
for idx, row in category_churn.iterrows():
    print(f"   - {row['계약대상_대분류명']}: {row['해약율']:.2%} ({row['계약대상_대분류_해약여부']:.0f}/{row['고객코드']:.0f})")

print(f"\n4. 다음 단계:")
print(f"   - 특성 엔지니어링 (Feature Engineering)")
print(f"   - 모델 학습 (XGBoost, Random Forest, LightGBM 등)")
print(f"   - 클래스 불균형 처리 (SMOTE, 가중치 조정 등)")
print(f"   - 모델 평가 (Precision, Recall, F1-Score, AUC-ROC)")
print("="*60)

# 시각화
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 해약여부 분포
ax1 = axes[0]
churn_counts = final_df_with_label['해약여부'].value_counts()
colors = ['#2ecc71', '#e74c3c']
ax1.bar([0, 1], churn_counts.values, color=colors, alpha=0.7, edgecolor='black')
ax1.set_xlabel('해약여부', fontsize=12, weight='bold')
ax1.set_ylabel('고객 수', fontsize=12, weight='bold')
ax1.set_title('해약여부 분포', fontsize=14, weight='bold')
ax1.set_xticks([0, 1])
ax1.set_xticklabels(['유지 (0)', '해약 (1)'])
for i, v in enumerate(churn_counts.values):
    ax1.text(i, v + 10, f'{v:,}명\n({v/len(final_df_with_label):.1%})', 
             ha='center', va='bottom', fontsize=11, weight='bold')

# 계약대상 대분류별 해약율
ax2 = axes[1]
category_churn_plot = category_churn.head(10)
bars = ax2.barh(category_churn_plot['계약대상_대분류명'], category_churn_plot['해약율'], 
                color='skyblue', alpha=0.7, edgecolor='black')
ax2.set_xlabel('해약율', fontsize=12, weight='bold')
ax2.set_ylabel('계약대상 대분류', fontsize=12, weight='bold')
ax2.set_title('계약대상 대분류별 해약율 (Top 10)', fontsize=14, weight='bold')
ax2.set_xlim(0, max(category_churn_plot['해약율']) * 1.2)
for i, (idx, row) in enumerate(category_churn_plot.iterrows()):
    ax2.text(row['해약율'] + 0.005, i, f"{row['해약율']:.1%}", 
             va='center', fontsize=10)

plt.tight_layout()
plt.show()
````

## 🔧 Feature Engineering: 상세 특성 생성

머신러닝 모델을 위한 상세 특성을 생성합니다.

### 특성 카테고리:
1. **계약 정보** (contracts_info) - 계약 건수, 금액, 기간 등
2. **고객 상호작용** (interaction_history) - 문의, 불만, AS 이력 등
3. **작업 이력** (작업이력) - 방문, 작업 빈도, 최근성 등
4. **지리적 특성** - 위치 기반 특성
5. **고객 기본 정보** - 업종, 규모, 등록일 등
6. **시간 기반 특성** - 고객 연령, 최근 활동 등

````python
# Step 2: Feature Engineering - 계약 정보 특성 (contracts_info)
import json
import numpy as np
from datetime import datetime

# 작업용 데이터프레임 복사
df_features = final_df_with_label.copy()

print("="*80)
print("🔧 Feature Engineering: 계약 정보 특성")
print("="*80)

# contracts_info에서 특성 추출하는 함수
def extract_contract_features(contracts_json):
    """계약 정보에서 특성 추출"""
    if pd.isna(contracts_json) or contracts_json == '':
        return {
            'contract_count': 0,
            'contract_total_area': 0,
            'contract_total_area_m': 0,
            'contract_avg_duration_days': 0,
            'contract_max_duration_days': 0,
            'contract_min_duration_days': 0,
            'contract_active_count': 0,
            'contract_cancelled_count': 0,
            'contract_unique_targets': 0,
            'contract_has_special_note': 0,
            'contract_days_since_first': np.nan,
            'contract_days_since_last': np.nan,
            'contract_대분류_count': 0,
            'contract_중분류_count': 0,
        }
    
    try:
        if isinstance(contracts_json, str):
            contracts = json.loads(contracts_json)
        else:
            contracts = contracts_json
            
        if not isinstance(contracts, list):
            contracts = []
    except:
        contracts = []
    
    if len(contracts) == 0:
        return {
            'contract_count': 0,
            'contract_total_area': 0,
            'contract_total_area_m': 0,
            'contract_avg_duration_days': 0,
            'contract_max_duration_days': 0,
            'contract_min_duration_days': 0,
            'contract_active_count': 0,
            'contract_cancelled_count': 0,
            'contract_unique_targets': 0,
            'contract_has_special_note': 0,
            'contract_days_since_first': np.nan,
            'contract_days_since_last': np.nan,
            'contract_대분류_count': 0,
            'contract_중분류_count': 0,
        }
    
    # 계약 건수
    contract_count = len(contracts)
    
    # 면적 합계
    total_area = sum([float(c.get('면적', 0) or 0) for c in contracts])
    total_area_m = sum([float(c.get('면적M', 0) or 0) for c in contracts])
    
    # 계약 기간 계산
    durations = []
    start_dates = []
    end_dates = []
    
    for c in contracts:
        try:
            start = pd.to_datetime(c.get('시작일자'), errors='coerce')
            end = pd.to_datetime(c.get('종료일자'), errors='coerce')
            if pd.notna(start) and pd.notna(end):
                duration = (end - start).days
                durations.append(duration)
                start_dates.append(start)
                end_dates.append(end)
        except:
            pass
    
    avg_duration = np.mean(durations) if durations else 0
    max_duration = max(durations) if durations else 0
    min_duration = min(durations) if durations else 0
    
    # 최초/최근 계약일로부터 경과일수
    days_since_first = np.nan
    days_since_last = np.nan
    if start_dates:
        days_since_first = (pd.Timestamp.now() - min(start_dates)).days
        days_since_last = (pd.Timestamp.now() - max(start_dates)).days
    
    # 활성/해약 계약 수
    active_count = sum([1 for c in contracts if pd.isna(c.get('해약일자'))])
    cancelled_count = sum([1 for c in contracts if pd.notna(c.get('해약일자'))])
    
    # 고유 계약대상 수
    unique_targets = len(set([c.get('계약대상') for c in contracts if c.get('계약대상')]))
    
    # 특이사항 여부
    has_special_note = int(any([c.get('계약특이사항') for c in contracts]))
    
    # 계약대상 분류 개수
    대분류_set = set([c.get('계약대상_대분류명') for c in contracts if c.get('계약대상_대분류명')])
    중분류_set = set([c.get('계약대상_중분류명') for c in contracts if c.get('계약대상_중분류명')])
    
    return {
        'contract_count': contract_count,
        'contract_total_area': total_area,
        'contract_total_area_m': total_area_m,
        'contract_avg_duration_days': avg_duration,
        'contract_max_duration_days': max_duration,
        'contract_min_duration_days': min_duration,
        'contract_active_count': active_count,
        'contract_cancelled_count': cancelled_count,
        'contract_unique_targets': unique_targets,
        'contract_has_special_note': has_special_note,
        'contract_days_since_first': days_since_first,
        'contract_days_since_last': days_since_last,
        'contract_대분류_count': len(대분류_set),
        'contract_중분류_count': len(중분류_set),
    }

# 계약 정보 특성 추출
print("Extracting contract features...")
contract_features = df_features['contracts_info'].apply(extract_contract_features)
contract_features_df = pd.DataFrame(list(contract_features))

# 데이터프레임에 추가
for col in contract_features_df.columns:
    df_features[col] = contract_features_df[col]

print(f"✅ Created {len(contract_features_df.columns)} contract features")
print(f"\nContract features:")
for col in contract_features_df.columns:
    print(f"   - {col}")

print(f"\n기본 통계:")
print(contract_features_df.describe().T[['mean', 'std', 'min', 'max']])
````

````python
# Step 3: Feature Engineering - 고객 상호작용 특성 (interaction_history)

print("\n" + "="*80)
print("🔧 Feature Engineering: 고객 상호작용 특성")
print("="*80)

def extract_interaction_features(interaction_json):
    """고객 상호작용 이력에서 특성 추출"""
    if pd.isna(interaction_json) or interaction_json == '':
        return {
            'interaction_count': 0,
            'interaction_complaint_count': 0,
            'interaction_inquiry_count': 0,
            'interaction_as_count': 0,
            'interaction_avg_response_days': 0,
            'interaction_max_response_days': 0,
            'interaction_days_since_last': np.nan,
            'interaction_has_unresolved': 0,
            'interaction_satisfaction_avg': 0,
        }
    
    try:
        if isinstance(interaction_json, str):
            interactions = json.loads(interaction_json)
        else:
            interactions = interaction_json
            
        if not isinstance(interactions, list):
            interactions = []
    except:
        interactions = []
    
    if len(interactions) == 0:
        return {
            'interaction_count': 0,
            'interaction_complaint_count': 0,
            'interaction_inquiry_count': 0,
            'interaction_as_count': 0,
            'interaction_avg_response_days': 0,
            'interaction_max_response_days': 0,
            'interaction_days_since_last': np.nan,
            'interaction_has_unresolved': 0,
            'interaction_satisfaction_avg': 0,
        }
    
    interaction_count = len(interactions)
    
    # 유형별 카운트 (키가 다를 수 있으므로 안전하게 처리)
    complaint_count = sum([1 for i in interactions if '불만' in str(i.get('상담유형', '')) or '클레임' in str(i.get('상담유형', ''))])
    inquiry_count = sum([1 for i in interactions if '문의' in str(i.get('상담유형', ''))])
    as_count = sum([1 for i in interactions if 'AS' in str(i.get('상담유형', '')) or 'A/S' in str(i.get('상담유형', ''))])
    
    # 응답 시간 계산
    response_times = []
    interaction_dates = []
    
    for i in interactions:
        try:
            # 접수일과 처리일 차이 계산
            receipt_date = pd.to_datetime(i.get('접수일자') or i.get('등록일자'), errors='coerce')
            process_date = pd.to_datetime(i.get('처리일자') or i.get('완료일자'), errors='coerce')
            
            if pd.notna(receipt_date):
                interaction_dates.append(receipt_date)
                
            if pd.notna(receipt_date) and pd.notna(process_date):
                response_time = (process_date - receipt_date).days
                if response_time >= 0:  # 음수 제외
                    response_times.append(response_time)
        except:
            pass
    
    avg_response_days = np.mean(response_times) if response_times else 0
    max_response_days = max(response_times) if response_times else 0
    
    # 최근 상호작용으로부터 경과일수
    days_since_last = np.nan
    if interaction_dates:
        days_since_last = (pd.Timestamp.now() - max(interaction_dates)).days
    
    # 미해결 건 여부
    has_unresolved = int(any([i.get('처리상태') == '미처리' or i.get('완료여부') == 'N' for i in interactions]))
    
    # 만족도 평균 (있는 경우)
    satisfaction_scores = []
    for i in interactions:
        try:
            score = i.get('만족도') or i.get('평점')
            if score and str(score).replace('.', '').isdigit():
                satisfaction_scores.append(float(score))
        except:
            pass
    
    satisfaction_avg = np.mean(satisfaction_scores) if satisfaction_scores else 0
    
    return {
        'interaction_count': interaction_count,
        'interaction_complaint_count': complaint_count,
        'interaction_inquiry_count': inquiry_count,
        'interaction_as_count': as_count,
        'interaction_avg_response_days': avg_response_days,
        'interaction_max_response_days': max_response_days,
        'interaction_days_since_last': days_since_last,
        'interaction_has_unresolved': has_unresolved,
        'interaction_satisfaction_avg': satisfaction_avg,
    }

# 상호작용 특성 추출
print("Extracting interaction features...")
interaction_features = df_features['interaction_history'].apply(extract_interaction_features)
interaction_features_df = pd.DataFrame(list(interaction_features))

# 데이터프레임에 추가
for col in interaction_features_df.columns:
    df_features[col] = interaction_features_df[col]

print(f"✅ Created {len(interaction_features_df.columns)} interaction features")
print(f"\nInteraction features:")
for col in interaction_features_df.columns:
    print(f"   - {col}")

print(f"\n기본 통계:")
print(interaction_features_df.describe().T[['mean', 'std', 'min', 'max']])
````

````python
# Step 4: Feature Engineering - 작업 이력 특성 (작업이력)

print("\n" + "="*80)
print("🔧 Feature Engineering: 작업 이력 특성")
print("="*80)

def extract_work_features(work_json):
    """작업 이력에서 특성 추출"""
    if pd.isna(work_json) or work_json == '':
        return {
            'work_count': 0,
            'work_confirmed_count': 0,
            'work_cancelled_count': 0,
            'work_unique_types': 0,
            'work_days_since_last': np.nan,
            'work_days_since_first': np.nan,
            'work_frequency_days': 0,  # 평균 작업 빈도 (일)
            'work_avg_csi_score': 0,
            'work_has_survey': 0,
            'work_pest_detected_count': 0,
            'work_avg_work_duration_min': 0,
            'work_total_divisions': 0,  # 총 구획 수
        }
    
    try:
        if isinstance(work_json, str):
            works = json.loads(work_json)
        else:
            works = work_json
            
        if not isinstance(works, list):
            works = []
    except:
        works = []
    
    if len(works) == 0:
        return {
            'work_count': 0,
            'work_confirmed_count': 0,
            'work_cancelled_count': 0,
            'work_unique_types': 0,
            'work_days_since_last': np.nan,
            'work_days_since_first': np.nan,
            'work_frequency_days': 0,
            'work_avg_csi_score': 0,
            'work_has_survey': 0,
            'work_pest_detected_count': 0,
            'work_avg_work_duration_min': 0,
            'work_total_divisions': 0,
        }
    
    work_count = len(works)
    
    # 확정/취소 작업 수
    confirmed_count = sum([1 for w in works if w.get('확정여부') == 'Y'])
    cancelled_count = sum([1 for w in works if w.get('취소여부') == 'Y'])
    
    # 고유 작업 유형 수
    unique_types = len(set([w.get('작업유형') for w in works if w.get('작업유형')]))
    
    # 작업 일자 분석
    work_dates = []
    for w in works:
        try:
            work_date = pd.to_datetime(w.get('작업일자'), errors='coerce')
            if pd.notna(work_date):
                work_dates.append(work_date)
        except:
            pass
    
    days_since_last = np.nan
    days_since_first = np.nan
    work_frequency_days = 0
    
    if work_dates:
        work_dates_sorted = sorted(work_dates)
        days_since_last = (pd.Timestamp.now() - work_dates_sorted[-1]).days
        days_since_first = (pd.Timestamp.now() - work_dates_sorted[0]).days
        
        # 평균 작업 빈도 계산
        if len(work_dates_sorted) > 1:
            total_span = (work_dates_sorted[-1] - work_dates_sorted[0]).days
            work_frequency_days = total_span / (len(work_dates_sorted) - 1)
    
    # CSI 점수 (서비스 만족도)
    csi_scores = []
    has_survey = 0
    
    for w in works:
        try:
            survey_data = w.get('서비스_만족도')
            if survey_data:
                has_survey = 1
                if isinstance(survey_data, str):
                    survey_list = json.loads(survey_data)
                else:
                    survey_list = survey_data
                    
                if isinstance(survey_list, list):
                    for s in survey_list:
                        score = s.get('평균_CSI_점수')
                        if score and str(score).replace('.', '').replace('-', '').isdigit():
                            csi_scores.append(float(score))
        except:
            pass
    
    avg_csi_score = np.mean(csi_scores) if csi_scores else 0
    
    # 해충 탐지 정보
    pest_detected_count = 0
    work_durations = []
    total_divisions = 0
    
    for w in works:
        try:
            # 작업 내역에서 개인 작업 분석
            work_details = w.get('작업내역')
            if work_details:
                if isinstance(work_details, str):
                    work_details_list = json.loads(work_details)
                else:
                    work_details_list = work_details
                
                if isinstance(work_details_list, list):
                    for wd in work_details_list:
                        # 작업 시간 계산
                        try:
                            fr_time = wd.get('서비스시작시간')
                            to_time = wd.get('서비스종료시간')
                            if fr_time and to_time:
                                fr = pd.to_datetime(fr_time, format='%H%M', errors='coerce')
                                to = pd.to_datetime(to_time, format='%H%M', errors='coerce')
                                if pd.notna(fr) and pd.notna(to):
                                    duration = (to - fr).total_seconds() / 60
                                    if duration > 0:
                                        work_durations.append(duration)
                        except:
                            pass
                        
                        # 개인작업상세에서 해충 정보
                        work_detail = wd.get('개인작업상세')
                        if work_detail:
                            if isinstance(work_detail, str):
                                detail_list = json.loads(work_detail)
                            else:
                                detail_list = work_detail
                            
                            if isinstance(detail_list, list):
                                for d in detail_list:
                                    # 해충이 발견된 경우
                                    pest_cnt = d.get('개체수') or d.get('해충마리수') or 0
                                    if pest_cnt and int(pest_cnt) > 0:
                                        pest_detected_count += 1
                                    
                                    # 구획 수
                                    if d.get('구획번호'):
                                        total_divisions += 1
        except:
            pass
    
    avg_work_duration = np.mean(work_durations) if work_durations else 0
    
    return {
        'work_count': work_count,
        'work_confirmed_count': confirmed_count,
        'work_cancelled_count': cancelled_count,
        'work_unique_types': unique_types,
        'work_days_since_last': days_since_last,
        'work_days_since_first': days_since_first,
        'work_frequency_days': work_frequency_days,
        'work_avg_csi_score': avg_csi_score,
        'work_has_survey': has_survey,
        'work_pest_detected_count': pest_detected_count,
        'work_avg_work_duration_min': avg_work_duration,
        'work_total_divisions': total_divisions,
    }

# 작업 이력 특성 추출
print("Extracting work history features...")
work_features = df_features['작업이력'].apply(extract_work_features)
work_features_df = pd.DataFrame(list(work_features))

# 데이터프레임에 추가
for col in work_features_df.columns:
    df_features[col] = work_features_df[col]

print(f"✅ Created {len(work_features_df.columns)} work history features")
print(f"\nWork history features:")
for col in work_features_df.columns:
    print(f"   - {col}")

print(f"\n기본 통계:")
print(work_features_df.describe().T[['mean', 'std', 'min', 'max']])
````

````python
# Step 5: Feature Engineering - 기본 고객 정보 및 파생 특성

print("\n" + "="*80)
print("🔧 Feature Engineering: 기본 정보 및 파생 특성")
print("="*80)

# 1. 고객 등록일로부터 경과일수 (고객 연령)
df_features['customer_age_days'] = (pd.Timestamp.now() - pd.to_datetime(df_features['등록일자'], errors='coerce')).dt.days

# 2. 지리적 특성
df_features['has_geo_info'] = ((df_features['위도'].notna()) & (df_features['경도'].notna())).astype(int)
df_features['has_utm_info'] = ((df_features['UTMK_X'].notna()) & (df_features['UTMK_Y'].notna())).astype(int)

# 3. 주소 정보
df_features['has_postal_code'] = df_features['우편번호'].notna().astype(int)
df_features['address_length'] = df_features['주소1'].fillna('').str.len()

# 4. 사업자 정보
df_features['has_business_no'] = df_features['사업자번호'].notna().astype(int)
df_features['업태_length'] = df_features['업태'].fillna('').str.len()
df_features['종목_length'] = df_features['종목'].fillna('').str.len()

# 5. 미수 내역
df_features['has_미수'] = df_features['미수내역'].notna().astype(int)

# 6. 계약 관련 파생 특성
df_features['contract_active_ratio'] = np.where(
    df_features['contract_count'] > 0,
    df_features['contract_active_count'] / df_features['contract_count'],
    0
)

df_features['contract_cancelled_ratio'] = np.where(
    df_features['contract_count'] > 0,
    df_features['contract_cancelled_count'] / df_features['contract_count'],
    0
)

# 7. 작업 빈도 관련 특성
df_features['work_per_contract'] = np.where(
    df_features['contract_count'] > 0,
    df_features['work_count'] / df_features['contract_count'],
    0
)

# 8. 최근성 지표 (Recency)
df_features['recency_work_vs_contract'] = df_features['work_days_since_last'] - df_features['contract_days_since_last']

# 9. 고객 활동성 점수
df_features['customer_activity_score'] = (
    (df_features['contract_count'] > 0).astype(int) +
    (df_features['work_count'] > 0).astype(int) +
    (df_features['interaction_count'] > 0).astype(int) +
    (df_features['work_has_survey'] == 1).astype(int)
)

# 10. 계약 다양성 지표
df_features['contract_diversity'] = (
    df_features['contract_대분류_count'] * 2 + 
    df_features['contract_중분류_count']
)

print(f"✅ Created additional derived features")

# 새로 생성된 특성 리스트
new_features = [
    'customer_age_days', 'has_geo_info', 'has_utm_info', 'has_postal_code',
    'address_length', 'has_business_no', '업태_length', '종목_length',
    'has_미수', 'contract_active_ratio', 'contract_cancelled_ratio',
    'work_per_contract', 'recency_work_vs_contract', 'customer_activity_score',
    'contract_diversity'
]

print(f"\nDerived features:")
for feat in new_features:
    print(f"   - {feat}")

print(f"\n기본 통계:")
print(df_features[new_features].describe().T[['mean', 'std', 'min', 'max']])
````

````python
# Step 6: Categorical Features - Label Encoding

print("\n" + "="*80)
print("🔧 Feature Engineering: 범주형 변수 인코딩")
print("="*80)

from sklearn.preprocessing import LabelEncoder

# 인코딩할 범주형 변수
categorical_features = {
    '유형대': 'customer_type_large_encoded',
    '유형중': 'customer_type_medium_encoded',
    '유형대_명칭': 'customer_type_large_name_encoded',
    '유형중_명칭': 'customer_type_medium_name_encoded',
    '대분류': 'industry_large_encoded',
    '중분류': 'industry_medium_encoded',
    '소분류': 'industry_small_encoded',
    '성별': 'gender_encoded',
    '국적': 'nationality_encoded',
    '신고객분류코드': 'new_customer_class_encoded',
}

label_encoders = {}

for original_col, encoded_col in categorical_features.items():
    if original_col in df_features.columns:
        # 결측값을 'UNKNOWN'으로 처리
        df_features[f'{original_col}_filled'] = df_features[original_col].fillna('UNKNOWN').astype(str)
        
        # Label Encoding
        le = LabelEncoder()
        df_features[encoded_col] = le.fit_transform(df_features[f'{original_col}_filled'])
        label_encoders[original_col] = le
        
        # 임시 컬럼 제거
        df_features.drop(f'{original_col}_filled', axis=1, inplace=True)
        
        print(f"   - {original_col} → {encoded_col} ({len(le.classes_)} classes)")

print(f"\n✅ Encoded {len(categorical_features)} categorical features")

# 범주형 변수의 고유값 수
print(f"\n범주형 변수 고유값 수:")
for original_col, encoded_col in categorical_features.items():
    if encoded_col in df_features.columns:
        n_unique = df_features[encoded_col].nunique()
        print(f"   - {encoded_col}: {n_unique} unique values")
````

````python
# Fix interaction_days_since_last by using the correct date field name

def extract_interaction_features_fixed(interaction_json):
    """Extract features from interaction_history JSON with corrected date field"""
    from datetime import datetime
    
    features = {}
    
    if pd.isna(interaction_json) or interaction_json == '' or interaction_json == '[]':
        features['interaction_count'] = 0
        features['interaction_voc_count'] = 0
        features['interaction_상담_count'] = 0
        features['interaction_민원_count'] = 0
        features['interaction_type_unique'] = 0
        features['interaction_avg_length'] = 0
        features['interaction_has_접수내용'] = 0
        features['interaction_recent_year'] = None
        features['interaction_days_since_last'] = None  # Will be filled
        return features
    
    try:
        # Parse JSON
        if isinstance(interaction_json, str):
            interactions = json.loads(interaction_json)
        else:
            interactions = interaction_json
            
        if not isinstance(interactions, list) or len(interactions) == 0:
            features['interaction_count'] = 0
            features['interaction_voc_count'] = 0
            features['interaction_상담_count'] = 0
            features['interaction_민원_count'] = 0
            features['interaction_type_unique'] = 0
            features['interaction_avg_length'] = 0
            features['interaction_has_접수내용'] = 0
            features['interaction_recent_year'] = None
            features['interaction_days_since_last'] = None
            return features
        
        # Count features
        features['interaction_count'] = len(interactions)
        
        # Event type counts
        event_types = [item.get('event_type', '') for item in interactions]
        features['interaction_voc_count'] = event_types.count('VOC')
        features['interaction_상담_count'] = event_types.count('상담')
        features['interaction_민원_count'] = event_types.count('민원')
        features['interaction_type_unique'] = len(set(event_types))
        
        # Content features
        contents = [item.get('접수내용', '') for item in interactions]
        valid_contents = [c for c in contents if c and c.strip()]
        features['interaction_avg_length'] = np.mean([len(str(c)) for c in valid_contents]) if valid_contents else 0
        features['interaction_has_접수내용'] = 1 if valid_contents else 0
        
        # Date features - using '접수일시' field
        dates = []
        for item in interactions:
            date_str = item.get('접수일시', '')  # Korean field name!
            if date_str:
                try:
                    # Parse ISO format: 2021-12-10T00:00:00
                    if 'T' in date_str:
                        date_obj = datetime.fromisoformat(date_str)
                    else:
                        date_obj = datetime.strptime(date_str, '%Y-%m-%d')
                    dates.append(date_obj)
                except:
                    pass
        
        if dates:
            most_recent_date = max(dates)
            features['interaction_recent_year'] = most_recent_date.year
            days_since = (datetime.now() - most_recent_date).days
            features['interaction_days_since_last'] = days_since
        else:
            features['interaction_recent_year'] = None
            features['interaction_days_since_last'] = None
            
    except Exception as e:
        # Return default values on error
        features['interaction_count'] = 0
        features['interaction_voc_count'] = 0
        features['interaction_상담_count'] = 0
        features['interaction_민원_count'] = 0
        features['interaction_type_unique'] = 0
        features['interaction_avg_length'] = 0
        features['interaction_has_접수내용'] = 0
        features['interaction_recent_year'] = None
        features['interaction_days_since_last'] = None
    
    return features

print("🔧 Fixed interaction feature extraction function")
print("   - Now correctly parsing '접수일시' field (Korean field name)")
print("   - This should fill interaction_days_since_last properly")
````

````python
# Create additional derived features for churn prediction

print("="*80)
print("🚀 Creating Additional Derived Features")
print("="*80)

original_feature_count = len(ml_dataset.columns)

# ============================================================================
# 1. RATIO & COMPARISON FEATURES
# ============================================================================
print("\n1️⃣ Ratio & Comparison Features")

# Interaction to work ratio (engagement balance)
ml_dataset['interaction_to_work_ratio'] = ml_dataset['interaction_count'] / (ml_dataset['work_count'] + 1)
print("   ✓ interaction_to_work_ratio: Balance between complaints and service visits")

# VOC intensity (complaint rate)
ml_dataset['voc_intensity'] = ml_dataset['interaction_voc_count'] / (ml_dataset['interaction_count'] + 1)
print("   ✓ voc_intensity: Proportion of VOC among all interactions")

# Work cancellation rate (service unreliability indicator)
ml_dataset['work_cancellation_rate'] = ml_dataset['work_cancelled_count'] / (ml_dataset['work_count'] + 1)
print("   ✓ work_cancellation_rate: Proportion of cancelled service visits")

# Contract cancellation rate
ml_dataset['contract_cancellation_rate'] = ml_dataset['contract_cancelled_count'] / (ml_dataset['contract_count'] + 1)
print("   ✓ contract_cancellation_rate: Historical cancellation tendency")

# Active contract ratio
ml_dataset['active_contract_ratio'] = ml_dataset['contract_active_count'] / (ml_dataset['contract_count'] + 1)
print("   ✓ active_contract_ratio: Proportion of currently active contracts")


# ============================================================================
# 2. LIFECYCLE & MATURITY FEATURES
# ============================================================================
print("\n2️⃣ Lifecycle & Maturity Features")

# Customer age in years
ml_dataset['customer_age_years'] = ml_dataset['contract_days_since_first'] / 365.25
print("   ✓ customer_age_years: How long customer has been with company")

# Service history length in years
ml_dataset['service_history_years'] = ml_dataset['work_days_since_first'] / 365.25
print("   ✓ service_history_years: Length of service relationship")

# Recent activity window (last 90 days indicator)
ml_dataset['recently_active_interaction'] = (ml_dataset['interaction_days_since_last'] <= 90).astype(int)
ml_dataset['recently_active_work'] = (ml_dataset['work_days_since_last'] <= 90).astype(int)
print("   ✓ recently_active_* : Binary indicators for activity in last 90 days")

# Dormant customer indicators
ml_dataset['is_dormant_interaction'] = (ml_dataset['interaction_days_since_last'] > 365).astype(int)
ml_dataset['is_dormant_work'] = (ml_dataset['work_days_since_last'] > 365).astype(int)
print("   ✓ is_dormant_* : Indicators for dormant customers (>1 year)")


# ============================================================================
# 3. ENGAGEMENT INTENSITY FEATURES
# ============================================================================
print("\n3️⃣ Engagement Intensity Features")

# Average interactions per year
ml_dataset['interaction_per_year'] = ml_dataset['interaction_count'] / (ml_dataset['customer_age_years'] + 0.1)
print("   ✓ interaction_per_year: Interaction frequency normalized by customer age")

# Average works per year
ml_dataset['work_per_year'] = ml_dataset['work_count'] / (ml_dataset['customer_age_years'] + 0.1)
print("   ✓ work_per_year: Service visit frequency normalized by customer age")

# Total engagement score (weighted combination)
ml_dataset['total_engagement_score'] = (
    ml_dataset['interaction_count'] * 0.3 +  # Interactions weighted less
    ml_dataset['work_count'] * 0.7  # Service visits weighted more
)
print("   ✓ total_engagement_score: Weighted combination of all engagements")

# Engagement diversity (unique types)
ml_dataset['engagement_diversity'] = (
    ml_dataset['interaction_type_unique'] + 
    ml_dataset['work_unique_types'] +
    ml_dataset['contract_unique_targets']
)
print("   ✓ engagement_diversity: Variety of engagement types")


# ============================================================================
# 4. SERVICE QUALITY INDICATORS
# ============================================================================
print("\n4️⃣ Service Quality Indicators")

# Work completion rate (already exists but let's ensure it)
if 'work_confirmed_count' in ml_dataset.columns:
    ml_dataset['work_completion_rate'] = ml_dataset['work_confirmed_count'] / (ml_dataset['work_count'] + 1)
    print("   ✓ work_completion_rate: Proportion of completed works")

# CSI reliability (has survey data)
if 'work_has_survey' in ml_dataset.columns and 'work_avg_csi_score' in ml_dataset.columns:
    ml_dataset['csi_reliability'] = ml_dataset['work_has_survey'].astype(int)
    ml_dataset['csi_quality'] = ml_dataset['work_avg_csi_score'].fillna(0)
    print("   ✓ csi_reliability, csi_quality: Customer satisfaction indicators")

# Pest detection rate (service effectiveness)
if 'work_pest_detected_count' in ml_dataset.columns:
    ml_dataset['pest_detection_rate'] = ml_dataset['work_pest_detected_count'] / (ml_dataset['work_count'] + 1)
    print("   ✓ pest_detection_rate: Proportion of visits that found pests")


# ============================================================================
# 5. RISK INDICATORS
# ============================================================================
print("\n5️⃣ Risk Indicators")

# Churn risk score based on recency (longer since last contact = higher risk)
ml_dataset['risk_score_recency'] = (
    ml_dataset['interaction_days_since_last'].fillna(365) / 365 * 0.4 +
    ml_dataset['work_days_since_last'].fillna(365) / 365 * 0.6
)
print("   ✓ risk_score_recency: Combined recency risk (0=active, higher=risky)")

# Churn risk score based on cancellations
ml_dataset['risk_score_cancellation'] = (
    ml_dataset['contract_cancellation_rate'] * 0.6 +
    ml_dataset['work_cancellation_rate'] * 0.4
)
print("   ✓ risk_score_cancellation: Historical cancellation tendency")

# Engagement decline indicator (comparing recent vs historical)
ml_dataset['engagement_decline'] = (
    (ml_dataset['interaction_days_since_last'] > ml_dataset['interaction_days_since_last'].median()).astype(int) +
    (ml_dataset['work_days_since_last'] > ml_dataset['work_days_since_last'].median()).astype(int)
) / 2
print("   ✓ engagement_decline: Indicator if customer is less active than average")

# Combined risk score
ml_dataset['combined_risk_score'] = (
    ml_dataset['risk_score_recency'] * 0.4 +
    ml_dataset['risk_score_cancellation'] * 0.3 +
    ml_dataset['engagement_decline'] * 0.3
)
print("   ✓ combined_risk_score: Comprehensive risk indicator")


# ============================================================================
# 6. TEMPORAL PATTERN FEATURES
# ============================================================================
print("\n6️⃣ Temporal Pattern Features")

# Recent year of activity
if 'interaction_recent_year' in ml_dataset.columns:
    ml_dataset['years_since_last_interaction'] = 2025 - ml_dataset['interaction_recent_year'].fillna(2000)
    print("   ✓ years_since_last_interaction: Years since most recent interaction")

# Service frequency (visits per unit time)
ml_dataset['service_frequency'] = ml_dataset['work_count'] / (ml_dataset['work_days_since_first'] / 30 + 1)
print("   ✓ service_frequency: Average works per month of service history")

# Interaction frequency
ml_dataset['interaction_frequency'] = ml_dataset['interaction_count'] / (ml_dataset['customer_age_years'] * 12 + 1)
print("   ✓ interaction_frequency: Average interactions per month")

# Time between services (average)
if 'work_frequency_days' in ml_dataset.columns:
    ml_dataset['avg_days_between_services'] = ml_dataset['work_frequency_days']
    print("   ✓ avg_days_between_services: Average gap between service visits")


# ============================================================================
# 7. CONTRACT-SPECIFIC FEATURES
# ============================================================================
print("\n7️⃣ Contract-Specific Features")

# Contract portfolio size (total managed area)
ml_dataset['portfolio_area_total'] = ml_dataset['contract_total_area'].fillna(0)
ml_dataset['portfolio_area_m_total'] = ml_dataset['contract_total_area_m'].fillna(0)
print("   ✓ portfolio_area_*: Total area under contract")

# Average contract area
ml_dataset['avg_contract_area'] = ml_dataset['contract_total_area'] / (ml_dataset['contract_count'] + 1)
ml_dataset['avg_contract_area_m'] = ml_dataset['contract_total_area_m'] / (ml_dataset['contract_count'] + 1)
print("   ✓ avg_contract_area_*: Average size per contract")

# Contract complexity (duration variance)
if 'contract_max_duration_days' in ml_dataset.columns and 'contract_min_duration_days' in ml_dataset.columns:
    ml_dataset['contract_duration_variance'] = (
        ml_dataset['contract_max_duration_days'] - ml_dataset['contract_min_duration_days']
    )
    print("   ✓ contract_duration_variance: Variety in contract lengths")

# Service intensity per contract
ml_dataset['services_per_contract'] = ml_dataset['work_count'] / (ml_dataset['contract_count'] + 1)
print("   ✓ services_per_contract: Average service visits per contract")


# ============================================================================
# 8. INTERACTION-SPECIFIC FEATURES
# ============================================================================
print("\n8️⃣ Interaction-Specific Features")

# Complaint intensity per service
ml_dataset['complaints_per_service'] = ml_dataset['interaction_voc_count'] / (ml_dataset['work_count'] + 1)
print("   ✓ complaints_per_service: VOC rate relative to service visits")

# Content engagement (has detailed content)
if 'interaction_has_접수내용' in ml_dataset.columns:
    ml_dataset['has_detailed_interactions'] = ml_dataset['interaction_has_접수내용']
    print("   ✓ has_detailed_interactions: Indicator for detailed interaction records")

# Average interaction content length
if 'interaction_avg_length' in ml_dataset.columns:
    ml_dataset['interaction_detail_level'] = ml_dataset['interaction_avg_length']
    print("   ✓ interaction_detail_level: Average content length of interactions")


# ============================================================================
# SUMMARY
# ============================================================================
new_feature_count = len(ml_dataset.columns)
added_features = new_feature_count - original_feature_count

print("\n" + "="*80)
print(f"✅ Feature Engineering Complete!")
print("="*80)
print(f"   Original features: {original_feature_count - 3}")  # Exclude ID and target
print(f"   New derived features: {added_features}")
print(f"   Total features now: {new_feature_count - 3}")  # Exclude ID and target
print(f"   Dataset shape: {ml_dataset.shape}")

# Check for any NaN values in new features
new_cols = ml_dataset.columns[original_feature_count:]
nan_counts = ml_dataset[new_cols].isnull().sum()
features_with_nans = nan_counts[nan_counts > 0]

if len(features_with_nans) > 0:
    print(f"\n⚠️ New features with missing values ({len(features_with_nans)}):")
    for feat, count in features_with_nans.items():
        print(f"   - {feat}: {count} ({count/len(ml_dataset)*100:.1f}%)")
else:
    print(f"\n✓ No missing values in new features")

print("="*80)
````

````python
# Add time-windowed features and remove combined_risk_score

print("="*80)
print("🔧 Adding Time-Windowed Features & Removing Combined Risk Score")
print("="*80)

# First, let's extract more detailed temporal information from JSON columns
# We need the actual dates from interaction_history and work_history

def extract_interaction_dates(interaction_json):
    """Extract all interaction dates for time window analysis"""
    from datetime import datetime
    
    if pd.isna(interaction_json) or interaction_json == '' or interaction_json == '[]':
        return []
    
    try:
        if isinstance(interaction_json, str):
            interactions = json.loads(interaction_json)
        else:
            interactions = interaction_json
            
        if not isinstance(interactions, list):
            return []
        
        dates = []
        for item in interactions:
            date_str = item.get('접수일시', '')
            if date_str:
                try:
                    if 'T' in date_str:
                        date_obj = datetime.fromisoformat(date_str)
                    else:
                        date_obj = datetime.strptime(date_str, '%Y-%m-%d')
                    dates.append(date_obj)
                except:
                    pass
        return dates
    except:
        return []

def extract_work_dates(work_json):
    """Extract all work dates for time window analysis"""
    from datetime import datetime
    
    if pd.isna(work_json) or work_json == '' or work_json == '[]':
        return []
    
    try:
        if isinstance(work_json, str):
            works = json.loads(work_json)
        else:
            works = work_json
            
        if not isinstance(works, list):
            return []
        
        dates = []
        for item in works:
            # Try multiple date field names
            date_str = item.get('작업일자', item.get('작업일시', item.get('date', '')))
            if date_str:
                try:
                    if 'T' in str(date_str):
                        date_obj = datetime.fromisoformat(str(date_str))
                    else:
                        date_obj = datetime.strptime(str(date_str), '%Y-%m-%d')
                    dates.append(date_obj)
                except:
                    pass
        return dates
    except:
        return []

print("\n1️⃣ Extracting interaction dates...")
interaction_dates = final_df_with_label['interaction_history'].apply(extract_interaction_dates)

print("2️⃣ Extracting work dates...")
work_dates = final_df_with_label['작업이력'].apply(extract_work_dates)

# Now create time-windowed features
from datetime import datetime, timedelta

reference_date = datetime.now()

print("\n3️⃣ Creating time-windowed features...")

# Interaction features for different time windows
time_windows = [30, 60, 90, 180, 365]

for days in time_windows:
    cutoff_date = reference_date - timedelta(days=days)
    
    # Count interactions in this window
    ml_dataset[f'interaction_last_{days}_days'] = interaction_dates.apply(
        lambda dates: sum(1 for d in dates if d >= cutoff_date)
    )
    
    # Count works in this window
    ml_dataset[f'work_last_{days}_days'] = work_dates.apply(
        lambda dates: sum(1 for d in dates if d >= cutoff_date)
    )
    
    print(f"   ✓ Created features for last {days} days")

# Create ratio features (recent vs total activity)
print("\n4️⃣ Creating activity ratio features...")

ml_dataset['interaction_ratio_30_to_total'] = ml_dataset['interaction_last_30_days'] / (ml_dataset['interaction_count'] + 1)
ml_dataset['interaction_ratio_90_to_total'] = ml_dataset['interaction_last_90_days'] / (ml_dataset['interaction_count'] + 1)
ml_dataset['work_ratio_30_to_total'] = ml_dataset['work_last_30_days'] / (ml_dataset['work_count'] + 1)
ml_dataset['work_ratio_90_to_total'] = ml_dataset['work_last_90_days'] / (ml_dataset['work_count'] + 1)

print("   ✓ interaction_ratio_30_to_total")
print("   ✓ interaction_ratio_90_to_total")
print("   ✓ work_ratio_30_to_total")
print("   ✓ work_ratio_90_to_total")

# Trend indicators (comparing recent windows)
print("\n5️⃣ Creating trend indicators...")

ml_dataset['interaction_trend_30_vs_60'] = ml_dataset['interaction_last_30_days'] / (ml_dataset['interaction_last_60_days'] - ml_dataset['interaction_last_30_days'] + 1)
ml_dataset['interaction_trend_60_vs_90'] = ml_dataset['interaction_last_60_days'] / (ml_dataset['interaction_last_90_days'] - ml_dataset['interaction_last_60_days'] + 1)
ml_dataset['work_trend_30_vs_60'] = ml_dataset['work_last_30_days'] / (ml_dataset['work_last_60_days'] - ml_dataset['work_last_30_days'] + 1)
ml_dataset['work_trend_60_vs_90'] = ml_dataset['work_last_60_days'] / (ml_dataset['work_last_90_days'] - ml_dataset['work_last_60_days'] + 1)

print("   ✓ interaction_trend_30_vs_60 (recent activity acceleration)")
print("   ✓ interaction_trend_60_vs_90")
print("   ✓ work_trend_30_vs_60")
print("   ✓ work_trend_60_vs_90")

# Activity decay indicators
print("\n6️⃣ Creating activity decay indicators...")

ml_dataset['interaction_decay_score'] = (
    ml_dataset['interaction_last_30_days'] * 1.0 +
    ml_dataset['interaction_last_60_days'] * 0.5 +
    ml_dataset['interaction_last_90_days'] * 0.25
) / (ml_dataset['interaction_count'] + 1)

ml_dataset['work_decay_score'] = (
    ml_dataset['work_last_30_days'] * 1.0 +
    ml_dataset['work_last_60_days'] * 0.5 +
    ml_dataset['work_last_90_days'] * 0.25
) / (ml_dataset['work_count'] + 1)

print("   ✓ interaction_decay_score (weighted recent activity)")
print("   ✓ work_decay_score")

# Remove combined_risk_score
print("\n7️⃣ Removing combined_risk_score...")
if 'combined_risk_score' in ml_dataset.columns:
    ml_dataset = ml_dataset.drop(columns=['combined_risk_score'])
    print("   ✓ Removed combined_risk_score")
else:
    print("   ⚠️ combined_risk_score not found in dataset")

# Summary
print("\n" + "="*80)
print("✅ Feature Engineering Update Complete!")
print("="*80)
print(f"Dataset shape: {ml_dataset.shape}")
print(f"\nAdded features:")
print(f"   - Time-windowed counts (10): interaction/work for 30/60/90/180/365 days")
print(f"   - Activity ratios (4): Recent activity vs total")
print(f"   - Trend indicators (4): Comparing adjacent time windows")
print(f"   - Decay scores (2): Weighted recent activity")
print(f"   Total new features: 20")
print(f"\nRemoved features:")
print(f"   - combined_risk_score")
print(f"\nNet change: +19 features")
print("="*80)
````

````python
# Fix work date extraction and recreate time-windowed features

def extract_work_dates_fixed(work_json):
    """Extract all work dates for time window analysis - FIXED"""
    from datetime import datetime
    
    if pd.isna(work_json) or work_json == '' or work_json == '[]':
        return []
    
    try:
        if isinstance(work_json, str):
            works = json.loads(work_json)
        else:
            works = work_json
            
        if not isinstance(works, list):
            return []
        
        dates = []
        for item in works:
            # The field is '작업일자' in YYYYMMDD format
            date_str = item.get('작업일자', '')
            if date_str:
                try:
                    # Parse YYYYMMDD format
                    date_obj = datetime.strptime(str(date_str), '%Y%m%d')
                    dates.append(date_obj)
                except:
                    pass
        return dates
    except:
        return []

print("="*80)
print("🔧 Fixing Work Date Extraction and Recreating Time-Windowed Features")
print("="*80)

print("\n1️⃣ Re-extracting work dates with correct field name...")
work_dates_fixed = final_df_with_label['작업이력'].apply(extract_work_dates_fixed)

# Verify extraction worked
total_work_dates = sum(len(dates) for dates in work_dates_fixed)
print(f"   ✓ Extracted {total_work_dates:,} work dates from {len(work_dates_fixed)} customers")
print(f"   ✓ Average: {total_work_dates / len(work_dates_fixed):.1f} work dates per customer")

# Recreate work time-windowed features
from datetime import datetime, timedelta

reference_date = datetime.now()

print("\n2️⃣ Recreating work time-windowed features...")

time_windows = [30, 60, 90, 180, 365]

for days in time_windows:
    cutoff_date = reference_date - timedelta(days=days)
    
    # Count works in this window
    ml_dataset[f'work_last_{days}_days'] = work_dates_fixed.apply(
        lambda dates: sum(1 for d in dates if d >= cutoff_date)
    )
    
    print(f"   ✓ Updated work_last_{days}_days")

# Recreate work-specific derived features
print("\n3️⃣ Recreating work ratio and trend features...")

ml_dataset['work_ratio_30_to_total'] = ml_dataset['work_last_30_days'] / (ml_dataset['work_count'] + 1)
ml_dataset['work_ratio_90_to_total'] = ml_dataset['work_last_90_days'] / (ml_dataset['work_count'] + 1)

ml_dataset['work_trend_30_vs_60'] = ml_dataset['work_last_30_days'] / (ml_dataset['work_last_60_days'] - ml_dataset['work_last_30_days'] + 1)
ml_dataset['work_trend_60_vs_90'] = ml_dataset['work_last_60_days'] / (ml_dataset['work_last_90_days'] - ml_dataset['work_last_60_days'] + 1)

ml_dataset['work_decay_score'] = (
    ml_dataset['work_last_30_days'] * 1.0 +
    ml_dataset['work_last_60_days'] * 0.5 +
    ml_dataset['work_last_90_days'] * 0.25
) / (ml_dataset['work_count'] + 1)

print("   ✓ work_ratio_30_to_total")
print("   ✓ work_ratio_90_to_total")
print("   ✓ work_trend_30_vs_60")
print("   ✓ work_trend_60_vs_90")
print("   ✓ work_decay_score")

# Show sample statistics
print("\n4️⃣ Sample Statistics:")
print(f"\n   Work Time Windows (All Customers):")
for days in time_windows:
    col = f'work_last_{days}_days'
    mean_val = ml_dataset[col].mean()
    median_val = ml_dataset[col].median()
    max_val = ml_dataset[col].max()
    print(f"   {col:25} - Mean: {mean_val:5.2f}, Median: {median_val:3.0f}, Max: {max_val:3.0f}")

print(f"\n   Comparison by Churn Status (Last 90 days):")
churned = ml_dataset[ml_dataset['해약여부'] == 1]
retained = ml_dataset[ml_dataset['해약여부'] == 0]

print(f"   - Churned:  {churned['work_last_90_days'].mean():.2f} (median: {churned['work_last_90_days'].median():.0f})")
print(f"   - Retained: {retained['work_last_90_days'].mean():.2f} (median: {retained['work_last_90_days'].median():.0f})")

print("\n" + "="*80)
print("✅ Work Time-Windowed Features Fixed!")
print("="*80)
````

## Step 1: Data Preprocessing

Prepare the data for modeling:
- Separate features from target
- Handle missing values strategically
- Split into train/validation/test sets with stratification
- Scale features

````python
# Separate features, target, and metadata
metadata_cols = ['고객코드', '고객명', '최근_해약일자_SM', '최초_해약일자_SM']
target_col = '해약여부'

feature_cols = [col for col in ml_dataset.columns 
                if col not in metadata_cols + [target_col]]

X = ml_dataset[feature_cols].copy()
y = ml_dataset[target_col].copy()
customer_ids = ml_dataset['고객코드'].copy()

print(f"✅ Data Separated:")
print(f"   Features (X): {X.shape}")
print(f"   Target (y): {y.shape}")
print(f"   Feature count: {len(feature_cols)}")
print(f"\n📊 Target Distribution:")
print(y.value_counts())
print(f"   Churn Rate: {y.mean()*100:.2f}%")
````

````python
# Analyze missing values
print("🔍 Missing Value Analysis:")
print("=" * 80)

missing_stats = pd.DataFrame({
    'Missing_Count': X.isnull().sum(),
    'Missing_Pct': (X.isnull().sum() / len(X) * 100).round(2)
}).sort_values('Missing_Pct', ascending=False)

# Show features with missing values
missing_features = missing_stats[missing_stats['Missing_Count'] > 0]
print(f"\nFeatures with missing values: {len(missing_features)} / {len(feature_cols)}")
print(f"\nTop 20 features with most missing data:")
print(missing_features.head(20))

# Categorize by missing percentage
high_missing = missing_features[missing_features['Missing_Pct'] > 50]
medium_missing = missing_features[(missing_features['Missing_Pct'] > 5) & (missing_features['Missing_Pct'] <= 50)]
low_missing = missing_features[missing_features['Missing_Pct'] <= 5]

print(f"\n📊 Missing Value Categories:")
print(f"   High (>50%): {len(high_missing)} features")
print(f"   Medium (5-50%): {len(medium_missing)} features")
print(f"   Low (<5%): {len(low_missing)} features")
````

````python
# Handle missing values strategically
from sklearn.impute import SimpleImputer

print("🔧 Handling Missing Values...")
print("=" * 80)

X_imputed = X.copy()

# Strategy 1: For recency features (days_since_last), use a high value (no recent activity)
recency_cols = [col for col in X.columns if 'days_since_last' in col]
print(f"\n1️⃣ Recency features (impute with 9999): {len(recency_cols)}")
for col in recency_cols:
    missing_count = X_imputed[col].isnull().sum()
    if missing_count > 0:
        X_imputed[col].fillna(9999, inplace=True)
        print(f"   ✓ {col}: {missing_count} nulls → 9999")

# Strategy 2: For count/frequency features, use 0 (no activity)
count_cols = [col for col in X.columns if any(word in col for word in ['count', 'frequency', 'ratio', 'last_30', 'last_60', 'last_90', 'last_180', 'last_365'])]
count_cols = [col for col in count_cols if col not in recency_cols]
print(f"\n2️⃣ Count/frequency features (impute with 0): {len(count_cols)}")
imputed_counts = 0
for col in count_cols:
    missing_count = X_imputed[col].isnull().sum()
    if missing_count > 0:
        X_imputed[col].fillna(0, inplace=True)
        imputed_counts += 1
print(f"   ✓ Imputed {imputed_counts} features with 0")

# Strategy 3: For score/quality features, use median
score_cols = [col for col in X.columns if any(word in col for word in ['score', 'csi', 'quality'])]
score_cols = [col for col in score_cols if col not in recency_cols + count_cols]
print(f"\n3️⃣ Score/quality features (impute with median): {len(score_cols)}")
for col in score_cols:
    missing_count = X_imputed[col].isnull().sum()
    if missing_count > 0:
        median_val = X_imputed[col].median()
        X_imputed[col].fillna(median_val, inplace=True)
        print(f"   ✓ {col}: {missing_count} nulls → {median_val:.2f}")

# Strategy 4: For remaining numeric features, use median
remaining_numeric = X_imputed.select_dtypes(include=[np.number]).columns
remaining_with_nulls = [col for col in remaining_numeric if X_imputed[col].isnull().sum() > 0]
print(f"\n4️⃣ Remaining numeric features (impute with median): {len(remaining_with_nulls)}")
if len(remaining_with_nulls) > 0:
    imputer = SimpleImputer(strategy='median')
    X_imputed[remaining_with_nulls] = imputer.fit_transform(X_imputed[remaining_with_nulls])
    print(f"   ✓ Imputed {len(remaining_with_nulls)} features")

# Verify no missing values remain
remaining_nulls = X_imputed.isnull().sum().sum()
print(f"\n✅ Missing values after imputation: {remaining_nulls}")

if remaining_nulls > 0:
    print(f"⚠️ Still have {remaining_nulls} nulls in:")
    print(X_imputed.isnull().sum()[X_imputed.isnull().sum() > 0])
````

````python
# Train/Validation/Test Split with Stratification
from sklearn.model_selection import train_test_split

print("📊 Splitting Data (Stratified)...")
print("=" * 80)

# First split: 70% train+val, 30% test
X_temp, X_test, y_temp, y_test, ids_temp, ids_test = train_test_split(
    X_imputed, y, customer_ids, 
    test_size=0.30, 
    random_state=42, 
    stratify=y
)

# Second split: 70% train (49% of total), 30% validation (21% of total)
X_train, X_val, y_train, y_val, ids_train, ids_val = train_test_split(
    X_temp, y_temp, ids_temp,
    test_size=0.30,
    random_state=42,
    stratify=y_temp
)

print(f"\n✅ Split Complete:")
print(f"   Train: {X_train.shape[0]:,} samples ({X_train.shape[0]/len(X)*100:.1f}%)")
print(f"   Val:   {X_val.shape[0]:,} samples ({X_val.shape[0]/len(X)*100:.1f}%)")
print(f"   Test:  {X_test.shape[0]:,} samples ({X_test.shape[0]/len(X)*100:.1f}%)")

print(f"\n📈 Churn Distribution (Stratification Check):")
print(f"   Train: {y_train.mean()*100:.2f}% churn ({y_train.sum()} / {len(y_train)})")
print(f"   Val:   {y_val.mean()*100:.2f}% churn ({y_val.sum()} / {len(y_val)})")
print(f"   Test:  {y_test.mean()*100:.2f}% churn ({y_test.sum()} / {len(y_test)})")
print(f"   Original: {y.mean()*100:.2f}% churn")

print(f"\n✅ Class distribution maintained across all splits!")
````

## Step 2: Baseline Model - XGBoost

Train an XGBoost classifier with class weights to handle imbalance:
- No feature scaling needed (tree-based model)
- Use `scale_pos_weight` for class imbalance
- Evaluate on validation set

````python
# Train XGBoost with class weight handling
from sklearn.metrics import (classification_report, confusion_matrix, 
                             roc_auc_score, roc_curve, precision_recall_curve,
                             average_precision_score, f1_score)
import time

print("🚀 Training XGBoost Classifier...")
print("=" * 80)

# Calculate scale_pos_weight for class imbalance
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()
print(f"\n⚖️ Class Balance:")
print(f"   Negative samples: {(y_train == 0).sum()}")
print(f"   Positive samples: {(y_train == 1).sum()}")
print(f"   Scale pos weight: {scale_pos_weight:.2f}")

# XGBoost parameters
params = {
    'objective': 'binary:logistic',
    'eval_metric': ['logloss', 'auc'],
    'scale_pos_weight': scale_pos_weight,
    'max_depth': 6,
    'learning_rate': 0.1,
    'n_estimators': 100,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'random_state': 42,
    'tree_method': 'hist',
    'early_stopping_rounds': 10
}

print(f"\n📝 Model Parameters:")
for key, value in params.items():
    print(f"   {key}: {value}")

# Train model
start_time = time.time()
model = xgb.XGBClassifier(**params)

model.fit(
    X_train, y_train,
    eval_set=[(X_train, y_train), (X_val, y_val)],
    verbose=True
)

training_time = time.time() - start_time
print(f"\n✅ Training Complete!")
print(f"   Time: {training_time:.2f} seconds")
print(f"   Best iteration: {model.best_iteration}")
print(f"   Best score: {model.best_score:.4f}")
````

````python
# Make predictions on validation set
print("🎯 Making Predictions on Validation Set...")
print("=" * 80)

y_val_pred_proba = model.predict_proba(X_val)[:, 1]
y_val_pred = model.predict(X_val)

# Calculate metrics
val_auc_roc = roc_auc_score(y_val, y_val_pred_proba)
val_avg_precision = average_precision_score(y_val, y_val_pred_proba)
val_f1 = f1_score(y_val, y_val_pred)

print(f"\n📊 Validation Set Performance:")
print(f"   AUC-ROC:          {val_auc_roc:.4f}")
print(f"   Average Precision: {val_avg_precision:.4f}")
print(f"   F1-Score:         {val_f1:.4f}")

print(f"\n📋 Classification Report:")
print(classification_report(y_val, y_val_pred, target_names=['Retained', 'Churned']))

print(f"\n🔍 Confusion Matrix:")
cm = confusion_matrix(y_val, y_val_pred)
print(f"                Predicted")
print(f"              Retained  Churned")
print(f"Actual Retained    {cm[0,0]:3d}      {cm[0,1]:3d}")
print(f"       Churned     {cm[1,0]:3d}      {cm[1,1]:3d}")

tn, fp, fn, tp = cm.ravel()
print(f"\n📈 Detailed Metrics:")
print(f"   True Negatives:  {tn} (correctly predicted retained)")
print(f"   False Positives: {fp} (predicted churn but retained)")
print(f"   False Negatives: {fn} (predicted retained but churned)")
print(f"   True Positives:  {tp} (correctly predicted churn)")
print(f"\n   Specificity: {tn/(tn+fp):.4f} (correctly identify retained)")
print(f"   Sensitivity: {tp/(tp+fn):.4f} (correctly identify churned)")
````

````python
# Visualize ROC and Precision-Recall Curves
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# ROC Curve
fpr, tpr, thresholds_roc = roc_curve(y_val, y_val_pred_proba)
ax1.plot(fpr, tpr, linewidth=2, label=f'XGBoost (AUC = {val_auc_roc:.4f})')
ax1.plot([0, 1], [0, 1], 'k--', linewidth=1, label='Random Classifier')
ax1.set_xlabel('False Positive Rate', fontsize=12)
ax1.set_ylabel('True Positive Rate', fontsize=12)
ax1.set_title('ROC Curve - Validation Set', fontsize=14, fontweight='bold')
ax1.legend(fontsize=10)
ax1.grid(alpha=0.3)

# Precision-Recall Curve
precision, recall, thresholds_pr = precision_recall_curve(y_val, y_val_pred_proba)
ax2.plot(recall, precision, linewidth=2, label=f'XGBoost (AP = {val_avg_precision:.4f})')
ax2.axhline(y=y_val.mean(), color='k', linestyle='--', linewidth=1, 
            label=f'Baseline ({y_val.mean():.4f})')
ax2.set_xlabel('Recall', fontsize=12)
ax2.set_ylabel('Precision', fontsize=12)
ax2.set_title('Precision-Recall Curve - Validation Set', fontsize=14, fontweight='bold')
ax2.legend(fontsize=10)
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("✅ ROC-AUC shows how well the model distinguishes between classes")
print("✅ PR-AUC is better for imbalanced datasets (focuses on positive class)")
````
